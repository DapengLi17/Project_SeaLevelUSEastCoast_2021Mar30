```python
lat1, lat2 = 20, 65
lon1, lon2 = -90+360, 360 # 90W

indir = ('/scratch/user/dapengli/Projects4iHESP/Project_SeaLevelUSEastCoast_2021Mar30/'
            'data_after_manipulation/UVVEL/SunwayMonthlyTransientHR/original/')
infile1_u = (indir + 
'B.E.13.BHISTC5.ne120_t12.sehires38.003.sunway.pop.h.1950-2005_UVEL.regrid_2021Apr07.nc')
infile1_v = (indir + 
'B.E.13.BHISTC5.ne120_t12.sehires38.003.sunway.pop.h.1950-2005_VVEL.regrid_2021Apr07.nc')
infile2_u = (indir + 
'2053_2100/B.E.13.BRCP85C5CN.ne120_t12.sehires38.003.sunway.CN_OFF.pop.h.2053-2100_UVEL.regrid_2021Apr07.nc')
infile2_v = (indir + 
'2053_2100/B.E.13.BRCP85C5CN.ne120_t12.sehires38.003.sunway.CN_OFF.pop.h.2053-2100_VVEL.regrid_2021Apr07.nc')
# infile2 = (indir + 
# 'B.E.13.BRCP85C5CN.ne30g16.sehires38.003.sunway.CN_OFF.pop.h.200601-210511_UVELVVEL.regrid_2021Apr06.nc')
ds1_u = xr.open_dataset(infile1_u)
ds1_v = xr.open_dataset(infile1_v)
ds2_u = xr.open_dataset(infile2_u)
ds2_v = xr.open_dataset(infile2_v)
# ds2 = xr.open_dataset(infile2)
ds1_u
```

```python
u1 = ds1_u.UVEL.isel(z_t=0).sel(time=slice('1991','2000'),lat=slice(lat1, lat2),lon=slice(lon1, lon2))
v1 = ds1_v.VVEL.isel(z_t=0).sel(time=slice('1991','2000'),lat=slice(lat1, lat2),lon=slice(lon1, lon2))
Vspd1 = xr.ufuncs.sqrt(u1.mean('time')**2 + v1.mean('time')**2)
u2 = ds2_u.UVEL.isel(z_t=0).sel(time=slice('2091','2100'),lat=slice(lat1, lat2),lon=slice(lon1, lon2))
v2 = ds2_v.VVEL.isel(z_t=0).sel(time=slice('2091','2100'),lat=slice(lat1, lat2),lon=slice(lon1, lon2))
v1
```

```python
du = u2.mean('time') - u1.mean('time')
dv = v2.mean('time') - v1.mean('time')
```

```python
fig = plt.figure(figsize=(28, 6))
pc = cartopy.crs.PlateCarree()
ax1 = plt.subplot(121,projection=pc)
ax1.coastlines()
# ax6.pcolormesh(dssh_rgd_HR.lon, dssh_rgd_HR.lat, dssh_rgd_HR, vmin=-20, vmax=20, transform=ccrs.PlateCarree(), cmap='bwr')
Q = ax1.quiver(u1.lon[0:-1:10], u1.lat[0:-1:10], u1.mean('time')[0:-1:10,0:-1:10], 
               v1.mean('time')[0:-1:10,0:-1:10], transform=pc)
ax1.set_title('1991-2001 av sfc vel (HR)')

ax2 = plt.subplot(122,projection=pc)
ax2.coastlines()
subplt2=ax2.pcolormesh(u1.lon, u1.lat, Vspd1, vmin=0, vmax=50,
               transform=pc)
# plot(vmax=50)
ax2.set_title('1991-2001 av sfc vel spd (HR)')

# # Now adding the colorbar
cb = fig.colorbar(subplt2, ax=[ax1,ax2], shrink=0.75,extend='max')
cb.set_label('[cm/s]')
```
