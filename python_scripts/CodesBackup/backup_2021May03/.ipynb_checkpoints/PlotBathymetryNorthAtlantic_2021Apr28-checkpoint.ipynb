# data analysis 

regrid t12 and g16 to 1deg grid for plots

```python
infile0_g16 = ('/scratch/group/ihesp/user/dapengli/Projects4iHESP/SupplementaryFiles4POP/'
'POPgrids_gx1v6_2020Jun29.nc') # POP g16 grid file
ds0_g16 = xr.open_dataset(infile0_g16)
ds0_g16
```

```python
infile0_t12 = ('/scratch/group/ihesp/user/dapengli/Projects4iHESP/SupplementaryFiles4POP/'
'POPgrids_tx0.1v2_2020Jun29.nc') # POP t12 grid file
ds0_t12 = xr.open_dataset(infile0_t12)
ds0_t12
```

```python
ds_1deg = xe.util.grid_global(1, 1)
ds_1deg
```

```python
ds0_g16_rn = ds0_g16.rename({'TLONG': 'lon', 'TLAT': 'lat'}) # rn: rename
ds0_t12_rn = ds0_t12.rename({'TLONG': 'lon', 'TLAT': 'lat'})
weightfile_t12_1deg = ('/scratch/group/ihesp/user/dapengli/Projects4iHESP/SupplementaryFiles4POP/'
                      'bilinear_t12_1x1degGlobalRectlinear_periodic_2020Aug01.nc')
regridder_t12_1deg = xe.Regridder(ds0_t12_rn, ds_1deg, method='bilinear', periodic=True,
                        filename=weightfile_t12_1deg, reuse_weights=True) # 
regridder_t12_1deg
```

```python
ds0_HR.HT.isel(nlat=2000).plot()
```

```python
HT_m_HR = ds0_HR.HT.where(ds0_HR.HT>0,np.nan) # m: masked, mask land as nan
HT_rgd_HR = regridder_t12_1deg(HT_m_HR)/100 # unit: m
HT_rgd_HR
```

```python
weightfile_g16_1deg = ('/scratch/group/ihesp/user/dapengli/Projects4iHESP/SupplementaryFiles4POP/'
                      'bilinear_g16_1x1degGlobalRectlinear_periodic_2020Aug01.nc')
regridder_g16_1deg = xe.Regridder(ds0_g16_rn, ds_1deg, method='bilinear', periodic=True,
                        filename=weightfile_g16_1deg, reuse_weights=True)
regridder_g16_1deg
```

```python
HT_m_LR = ds0_LR.HT.where(ds0_LR.HT>0,np.nan) # m: masked, mask land as nan
HT_rgd_LR = regridder_g16_1deg(HT_m_LR)/100 # unit: m
HT_rgd_LR
```

choose North Atlantic region

```python
ds_etopo_rn = ds_etopo.rename({'X': 'lon', 'Y': 'lat'}) # rn: rename
ds_etopo_rn
```

```python
regridder_obs5mindeg_1deg = xe.Regridder(ds_etopo_rn, ds_1deg, method='bilinear', periodic=True,
        filename='bilinear_obs0.08deg_1deg_Global_periodic_2021Apr28.nc', reuse_weights=True) # 
regridder_obs5mindeg_1deg
```

```python
bathy_etopo_rgd = regridder_obs5mindeg_1deg(bathy_etopo) # unit: m
bathy_etopo_rgd
```

```python
bathy_etopo_rgd.plot()
```

choose North Atlantic region

```python
HT_rgd_NA_LR = HT_rgd_LR.where((ds_1deg.lat>lat1) & (ds_1deg.lat<lat2) & 
                                   (ds_1deg.lon>lon1) & (ds_1deg.lon<lon2), drop=True)
HT_rgd_NA_HR = HT_rgd_HR.where((ds_1deg.lat>lat1) & (ds_1deg.lat<lat2) & 
                                   (ds_1deg.lon>lon1) &  (ds_1deg.lon<lon2), drop=True)
bathy_etopo_NA_rgd = bathy_etopo_rgd.where((ds_1deg.lat>lat1) & (ds_1deg.lat<lat2) & 
                                   (ds_1deg.lon>lon1) &  (ds_1deg.lon<lon2), drop=True)
HT_rgd_NA_HR
```

# make pics

```python
fig, axs = plt.subplots(2,3,figsize=(15,10))
(HT_rgd_NA_LR).plot(ax=axs[0,0],vmin=0,vmax=5000)
(HT_rgd_NA_HR).plot(ax=axs[0,1],vmin=0,vmax=5000)
bathy_etopo_NA_rgd.plot(ax=axs[0,2],vmin=0,vmax=5000)
(HT_rgd_NA_LR-bathy_etopo_NA_rgd).plot(ax=axs[1,0],vmin=-500,vmax=500,cmap='bwr')
(HT_rgd_NA_HR-bathy_etopo_NA_rgd).plot(ax=axs[1,1],vmin=-500,vmax=500,cmap='bwr')
((HT_rgd_NA_HR-HT_rgd_NA_LR)).plot(ax=axs[1,2],vmin=-500,vmax=500,cmap='bwr')
```

```python
pc = cartopy.crs.PlateCarree()

fig, axs = plot.subplots(nrows=2,ncols=3,proj='cyl',width='16in',height='6in')

cs1=axs[0,0].contour(HT_rgd_NA_LR.lon, HT_rgd_NA_LR.lat, 
HT_rgd_NA_LR, levels=[100], transform=pc, colors=['b'])
cs2=axs[0,0].contour(HT_rgd_NA_LR.lon, HT_rgd_NA_LR.lat, 
HT_rgd_NA_LR, levels=[1000,3000,5000], transform=pc,colors=['r','y','g'])
# axs[0,0].coastlines(resolution='110m', color='black', linewidth=1)
axs[0,0].set_extent([lon1, lon2, lat1, lat2], crs=pc)
axs[0,0].set_title('LR bathymetry')
axs[0,0].format(land=True, landcolor='grey')
axs[0,0].clabel(cs1, inline=False, fmt='%d', fontsize=10)
axs[0,0].clabel(cs2, inline=True, fmt='%d', fontsize=10)


cs1=axs[0,1].contour(HT_rgd_NA_HR.lon, HT_rgd_NA_HR.lat, 
HT_rgd_NA_HR, levels=[100], transform=pc, colors=['b'])
cs2=axs[0,1].contour(HT_rgd_NA_HR.lon, HT_rgd_NA_HR.lat, 
HT_rgd_NA_HR, levels=[1000,3000,5000], transform=pc,colors=['r','y','g'])
# axs[0,0].coastlines(resolution='110m', color='black', linewidth=1)
axs[0,1].set_extent([lon1, lon2, lat1, lat2], crs=pc)
axs[0,1].set_title('HR bathymetry')
axs[0,1].format(land=True, landcolor='grey')
axs[0,1].clabel(cs1, inline=False, fmt='%d', fontsize=10)
axs[0,1].clabel(cs2, inline=True, fmt='%d', fontsize=10)


cs1=axs[0,2].contour(bathy_etopo_NA_rgd.lon, bathy_etopo_NA_rgd.lat, 
bathy_etopo_NA_rgd, levels=[100], transform=pc, colors=['b'])
cs2=axs[0,2].contour(bathy_etopo_NA_rgd.lon, bathy_etopo_NA_rgd.lat, 
bathy_etopo_NA_rgd, levels=[1000,3000,5000], transform=pc,colors=['r','y','g'])
# axs[0,0].coastlines(resolution='110m', color='black', linewidth=1)
axs[0,2].set_extent([lon1, lon2, lat1, lat2], crs=pc)
axs[0,2].set_title('ETOPO5 bathymetry')
axs[0,2].format(land=True, landcolor='grey')
axs[0,2].clabel(cs1, inline=False, fmt='%d', fontsize=10)
axs[0,2].clabel(cs2, inline=True, fmt='%d', fontsize=10)


axs[1,0].contourf(HT_rgd_NA_LR.lon, HT_rgd_NA_LR.lat, 
(HT_rgd_NA_LR-bathy_etopo_NA_rgd), transform=pc, levels=np.arange(-500,500+1,200), 
                  vmin=-500,vmax=500,cmap='bwr', cmap_kw={'cut': -0.2},extend='both')
# axs[1,0].coastlines(resolution='110m', color='black', linewidth=1)
axs[1,0].set_extent([lon1, lon2, lat1, lat2], crs=pc)
axs[1,0].set_title('LR-ETOPO5 bathymetry')
axs[1,0].format(land=True, landcolor='grey')


axs[1,1].contourf(HT_rgd_NA_HR.lon, HT_rgd_NA_HR.lat, 
(HT_rgd_NA_HR-bathy_etopo_NA_rgd), transform=pc, levels=np.arange(-500,500+1,200), 
                  vmin=-500,vmax=500,cmap='bwr', cmap_kw={'cut': -0.2},extend='both')
# axs[1,1].coastlines(resolution='110m', color='black', linewidth=1)
axs[1,1].set_extent([lon1, lon2, lat1, lat2], crs=pc)
axs[1,1].set_title('HR-ETOPO5 bathymetry')
axs[1,1].format(land=True, landcolor='grey')


m2=axs[1,2].contourf(HT_rgd_NA_HR.lon, HT_rgd_NA_HR.lat, 
(HT_rgd_NA_HR-HT_rgd_NA_LR), transform=pc, levels=np.arange(-500,500+1,200), 
                  vmin=-500,vmax=500,cmap='bwr', cmap_kw={'cut': -0.2},extend='both')
# axs[1,2].coastlines(resolution='110m', color='black', linewidth=1)
axs[1,2].set_extent([lon1, lon2, lat1, lat2], crs=pc)
axs[1,2].set_title('HR-LR bathymetry')
axs[1,2].format(land=True, landcolor='grey')

cb = axs[1,2].colorbar(m2,loc='r', label='[m]', extend='both',width='2em',shrink=1)
```
