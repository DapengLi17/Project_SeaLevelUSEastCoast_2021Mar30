find the nearest lat and lon (Sunway HR)

```python
infile0_HR = ('/scratch/group/ihesp/archive/B.E.13.BHISTC5.ne120_t12.sehires38.003.sunway/ocn/monthly/'
              'B.E.13.BHISTC5.ne120_t12.sehires38.003.sunway.pop.h.1850-01.nc')
ds0_HR = xr.open_dataset(infile0_HR,chunks={'nlat':600,'nlon':600})
ds0_HR
```

```python
ds0_HR.sel(nlat=loc1_HR-3, nlon=loc2_HR).KMT.values
```

```python
ds0_HR.KMT.plot()
ds0_HR.sel(nlat=loc1_HR-2, nlon=loc2_HR).KMT.values
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,4))
ds0_HR.TLONG.plot(ax=axes[0])
ds0_HR.TLAT.plot(ax=axes[1])
```

```python
# A 2D plot of the SPEED variable, assigning the coordinate values,
# and plot the verticies of each point
# ds0_LR.KMT.plot()
plt.scatter(ds0_HR.TLONG, ds0_HR.TLAT)

# I want to find the speed at a certain lat/lon point.
lat,lon = 40.7, -74.0+360 # lat and lon of New York

# First, find the index of the grid point nearest a specific lat/lon.   
abslat = np.abs(ds0_HR.TLAT-lat)
abslon = np.abs(ds0_HR.TLONG-lon)
c = np.maximum(abslon, abslat)

([loc1_HR], [loc2_HR]) = np.where(c == np.min(c)) 
#DL loc1_HR, loc2_HR are the index for the 1st and 2nd dimensions
#DL 1st dimension is nlat and 2nd dimension is nlon

# Now I can use that index location to get the values at the x/y diminsion
point_ds = ds0_HR.sel(nlat=loc1_HR, nlon=loc2_HR)

# Plot requested lat/lon point blue
plt.scatter(lon, lat, color='b')
plt.text(lon, lat, 'requested')

# Plot nearest point in the array red
plt.scatter(point_ds.TLONG, point_ds.TLAT, color='r')
plt.text(point_ds.TLONG, point_ds.TLAT, 'nearest')

# plt.title('speed at nearest point: %s' % point_ds.SPEED.data)
print('index for the first and second dimension')
print(loc1_HR,loc2_HR)
```
