# lat and lon of selected tide gauge stations

```python
# lat_LB,lon_LB = 56.6, -61.7 # lat and lon of NAIN (newfoundland and labraddor), 
lat_HF, lon_HF = 44.7, -63.6 # lat and lon of HALIFAX (Nova Scotia)
lat_NY, lon_NY = 40.7, -74.0 # lat and lon of New York
lat_CH, lon_CH = 32.8, -79.9 # lat and lon of CHARLESTON I (South Carolina)
pic1 = '/scratch/user/dapengli/Projects4iHESP/Project_SeaLevelCO2_2021Mar26/Pics/TideGaugeStationsUSeastCoast_2021Mar28.png'
```

# make pics 

```python
fig = plt.figure(figsize=(12, 6))

pc = cartopy.crs.PlateCarree()
ax1 = plt.subplot(111,projection=pc)
ax1.coastlines()
# Plot tide gauge stations lat/lon point blue
ax1.scatter(lon_HF, lat_HF, color='b',transform=pc)
ax1.text(lon_HF+1, lat_HF-2, 'Halifax',transform=pc)
ax1.scatter(lon_NY, lat_NY, color='b',transform=pc)
ax1.text(lon_NY+1, lat_NY-2, 'New York',transform=pc)
ax1.scatter(lon_CH, lat_CH, color='b',transform=pc)
ax1.text(lon_CH+1, lat_CH-2, 'Charleston',transform=pc)
ax1.set_extent([-100, -20, 10, 75], crs=pc)
ax1.set_title('tide gauge stations')

ax1.set_xticks([-100, -80, -60, -40, -20], crs=pc)
ax1.set_yticks([20, 40, 60], crs=pc)

# plt.savefig(pic1,bbox_inches='tight')
```
