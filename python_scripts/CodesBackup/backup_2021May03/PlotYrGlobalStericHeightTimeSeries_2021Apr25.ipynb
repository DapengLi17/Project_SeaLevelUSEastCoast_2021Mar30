# data analysis

```python
h_rst_gl_av_TH = ds_TH.h_rst.weighted(ds0_HR.TAREA).mean(("nlon", "nlat"))
h_rst_gl_av_CH = ds_CH.h_rst.weighted(ds0_HR.TAREA).mean(("nlon", "nlat"))
h_rst_gl_av_TL = ds_TL.h_rst.weighted(ds0_LR.TAREA).mean(("nlon", "nlat"))

h_rst_gl_a_TH = h_rst_gl_av_TH - h_rst_gl_av_TH.isel(time=0) # a: anomaly
h_rst_gl_a_TH = h_rst_gl_a_TH.where(h_rst_gl_a_TH<50, np.nan) # mask spikes as nan
h_rst_gl_a_CH = h_rst_gl_av_CH - h_rst_gl_av_CH.isel(time=0)
h_rst_gl_a_TL = h_rst_gl_av_TL - h_rst_gl_av_TL.isel(time=0) # a: anomaly

h_rst_gl_a_TH
```

```python
fig, axs = plt.subplots(1,1,figsize=(12,4))
# ds_TH.h_rst.mean(('nlat','nlon')).plot(ax=axs,color='r')
# ds_CH.h_rst.mean(('nlat','nlon')).plot(ax=axs,color='b')
axs.plot(np.arange(1850,2100+1,1), h_rst_gl_a_TH, color='r', label='transient (HR)')
axs.plot(np.arange(1850,2100+1,1), h_rst_gl_a_CH, color='b', label='control (HR)')
axs.plot(np.arange(1850,2100+1,1), (h_rst_gl_a_TH.values-h_rst_gl_a_CH.values), color='k', 
          label='transient-control (HR)')
yr_no2014 = np.arange(1850,2100+1,1)
yr_no2014 = yr_no2014[yr_no2014 != 2014]
axs.plot(yr_no2014, h_rst_gl_a_TL, color='r', linestyle='dashed', label='transient (LR)')

axs.set_xlim([1845,2105])
axs.set_ylim([-10,40])
axs.set_ylabel('[cm]')
axs.set_title('global av steric height relative to 1850 (Sunway)')
axs.grid()
plt.legend(loc="upper left")
```
