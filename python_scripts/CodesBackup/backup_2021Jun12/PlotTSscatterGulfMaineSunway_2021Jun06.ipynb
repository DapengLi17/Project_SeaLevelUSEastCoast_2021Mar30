```python
%%time
# c: correct, indicating transient - control to correct for model drift 
temp1c_TH = (ds1_TH.TEMP.mean('time') - ds1_CH.TEMP.mean('time') + ds0_TH.TEMP.mean('time'))
temp1c_GM_TH = temp1c_TH.isel(nlon=slice(nlon1_t12_GM, nlon2_t12_GM),
                                        nlat=slice(nlat1_t12_GM, nlat2_t12_GM)).compute()

temp2c_TH = (ds2_TH.TEMP.mean('time') - ds2_CH.TEMP.mean('time') + ds0_TH.TEMP.mean('time'))
temp2c_GM_TH = temp2c_TH.isel(nlon=slice(nlon1_t12_GM, nlon2_t12_GM),
                                        nlat=slice(nlat1_t12_GM, nlat2_t12_GM)).compute()


temp1c_TL = (ds1_TL.TEMP.mean('time') - ds1_CL.TEMP.mean('time') + ds0_TL.TEMP.mean('time')) 
temp1c_GM_TL = temp1c_TL.isel(nlon=slice(nlon1_g16_GM, nlon2_g16_GM),
                                        nlat=slice(nlat1_g16_GM, nlat2_g16_GM)).compute()

temp2c_TL = (ds2_TL.TEMP.mean('time') - ds2_CL.TEMP.mean('time') + ds0_TL.TEMP.mean('time')) 
temp2c_GM_TL = temp2c_TL.isel(nlon=slice(nlon1_g16_GM, nlon2_g16_GM),
                                        nlat=slice(nlat1_g16_GM, nlat2_g16_GM)).compute()
```

```python
%%time
# c: correct, indicating transient - control to correct for model drift 
salt1c_TH = (ds1_TH.SALT.mean('time') - ds1_CH.SALT.mean('time') + ds0_TH.SALT.mean('time'))
salt1c_GM_TH = salt1c_TH.isel(nlon=slice(nlon1_t12_GM, nlon2_t12_GM),
                                        nlat=slice(nlat1_t12_GM, nlat2_t12_GM)).compute()
salt2c_TH = (ds2_TH.SALT.mean('time') - ds2_CH.SALT.mean('time') + ds0_TH.SALT.mean('time'))
salt2c_GM_TH = salt2c_TH.isel(nlon=slice(nlon1_t12_GM, nlon2_t12_GM),
                                        nlat=slice(nlat1_t12_GM, nlat2_t12_GM)).compute()

salt1c_TL = (ds1_TL.SALT.mean('time') - ds1_CL.SALT.mean('time') + ds0_TL.SALT.mean('time')) 
salt1c_GM_TL = salt1c_TL.isel(nlon=slice(nlon1_g16_GM, nlon2_g16_GM),
                                        nlat=slice(nlat1_g16_GM, nlat2_g16_GM)).compute()
salt2c_TL = (ds2_TL.SALT.mean('time') - ds2_CL.SALT.mean('time') + ds0_TL.SALT.mean('time')) 
salt2c_GM_TL = salt2c_TL.isel(nlon=slice(nlon1_g16_GM, nlon2_g16_GM),
                                        nlat=slice(nlat1_g16_GM, nlat2_g16_GM)).compute()
```

```python
%%time
# return RHO, dRHOdS, dRHOdT
# need to use dask for HR density calculation, otherwise there is a memory issue.
rho1_GM_TH, drhodS1_GM_TH, drhodT1_GM_TH = pop_tools.eos(salt1c_GM_TH, temp1c_GM_TH, 
                                                         depth=temp1c_GM_TH.z_t*1e-2, return_coefs=True)
rho2_GM_TH, drhodS2_GM_TH, drhodT2_GM_TH = pop_tools.eos(salt2c_GM_TH, temp2c_GM_TH, 
                                                         depth=temp2c_GM_TH.z_t*1e-2, return_coefs=True)

beta1_GM_TH = drhodS1_GM_TH/rho1_GM_TH/1000 
# https://github.com/NCAR/pop-tools/blob/master/pop_tools/eos.py DRHODS * 1000 (probably for unit convertion), 
# haline coefficient is ~ 1e-4 (e.g., Fig 1., Griffies et al., 2014 An assessment of global and regional sea level 
# for years 1993–2007 in a suite of interannual CORE-II simulations), so I divide 1000 to get the correct order of magnitude
beta2_GM_TH = drhodS2_GM_TH/rho2_GM_TH/1000
alpha1_GM_TH = drhodT1_GM_TH/rho1_GM_TH
alpha2_GM_TH = drhodT2_GM_TH/rho2_GM_TH

rho1_GM_TL, drhodS1_GM_TL, drhodT1_GM_TL = pop_tools.eos(salt1c_GM_TL, temp1c_GM_TL, 
                                                         depth=temp1c_GM_TL.z_t*1e-2, return_coefs=True)
rho2_GM_TL, drhodS2_GM_TL, drhodT2_GM_TL = pop_tools.eos(salt2c_GM_TL, temp2c_GM_TL, 
                                                         depth=temp2c_GM_TL.z_t*1e-2, return_coefs=True)

beta1_GM_TL = drhodS1_GM_TL/rho1_GM_TL/1000
beta2_GM_TL = drhodS2_GM_TL/rho2_GM_TL/1000
alpha1_GM_TL = drhodT1_GM_TL/rho1_GM_TL
alpha2_GM_TL = drhodT2_GM_TL/rho2_GM_TL

drhodS1_GM_TH
```
