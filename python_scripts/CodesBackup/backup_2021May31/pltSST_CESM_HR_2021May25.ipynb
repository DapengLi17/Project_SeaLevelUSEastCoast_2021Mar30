# import modules

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy
cartopy.config['pre_existing_data_dir']='/scratch/group/ihesp/shared/cartopy_features'
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
```

# plot HR CESM-POP

```python
infile_HR = ('/ihesp/archive/HRMIP/'
             'B.E.13.B1950TRC5.ne120_t12.cesm-ihesp-1950-2050.013/ocn/hist/'
             'B.E.13.B1950TRC5.ne120_t12.cesm-ihesp-1950-2050.013.pop.h.nday1.2050-09-01.nc')
ds_HR = xr.open_dataset(infile_HR)
ds_HR
```

```python
sst_HR = ds_HR.SST.isel(time=0)
sst_HR = sst_HR.where(ds_HR.KMT>0,np.nan) # use ds_HR.KMT to mask land with NaN
```

```python
# === input grids ===
infile0_t12 = ('/scratch/group/ihesp/user/dapengli/Projects4iHESP/SupplementaryFiles4POP/'
'POPgrids_tx0.1v2_2020Jun29.nc') # POP t12 grid file
ds_grid_tx01v2 = xr.open_dataset(infile0_t12)
ds_grid_tx01v2
```

# make pics

```python
proj = ccrs.Robinson(central_longitude=0)
pc = ccrs.PlateCarree()
lon_formatter = LongitudeFormatter(zero_direction_label=True)
lat_formatter = LatitudeFormatter()
fig = plt.figure(figsize=(15, 8))

axs1 = plt.axes(projection=ccrs.PlateCarree())
# ax1 = fig.add_subplot(111,projection=pc)
subplt1=axs1.pcolormesh(ds_grid_tx01v2.TLONG, ds_grid_tx01v2.TLAT, sst_HR,
                       vmin=-10,vmax=40,transform=pc,cmap='jet',shading='flat')
axs1.coastlines(resolution='110m', color='black')
axs1.set_extent([-180, 180, -90, 90], crs=pc)
axs1.set_xticks(np.arange(-180,180+1,60), crs=pc)
axs1.set_xticklabels(np.arange(-180,180+1,60), fontsize=12)
axs1.xaxis.set_major_formatter(lon_formatter)
axs1.set_yticks(np.arange(-90,90+1,30), crs=pc)
axs1.set_yticklabels(np.arange(-90,90+1,30), fontsize=12)
axs1.yaxis.set_major_formatter(lat_formatter)
# ax1.stock_img()
axs1.add_feature(cartopy.feature.LAND, zorder=0, edgecolor='black',facecolor='grey')
axs1.set_title('Global SST CESM-HR')
plt.colorbar(subplt1,shrink=0.8,label='[$^\circ$ C]')
fig.set_facecolor("w")
fig.savefig('SST_Global_CESM_HR_2021May25.png',bbox_inches='tight')
```
