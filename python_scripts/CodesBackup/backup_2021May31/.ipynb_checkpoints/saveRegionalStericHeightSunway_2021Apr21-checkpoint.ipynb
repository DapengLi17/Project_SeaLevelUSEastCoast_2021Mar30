Eq for steric height ($ h_{steric} $, Tomczak, Eq (2.3)a; Lombard et al., 2005 Eq(1)): <br>
$ h_{steric} = \int_{-H}^{0} \frac{\rho_0(x,y,z)-\rho(x,y,z,t)}{\rho_0(x,y,z)}dz  = \int_{-H}^{0} 1- \frac{\rho(x,y,z,t)}{\rho_0(x,y,z)}dz $ <br>
H: total water depth (Lombard et al., 2009), $\rho(x,y,z,t)$ density (Tomczak, Eq (2.3)a), 
$\rho_0 = 1028 kg/m^3$ reference density (see Eq (3), van Western et al., (2020 nature-scientific report))

extra notes: <br>
rho stands for in-situ density <br>
NCAR pop-tool.eos compute (in-situ) density with potential temperature, see https://pop-tools.readthedocs.io/en/latest/examples/eos.html <br>
vertical integration is essentially a vertical weighted (weight is dz) sum (see https://github.com/NCAR/notebook-gallery/blob/master/notebooks/Run-Anywhere/Ocean-Heat-Content/OHC_tutorial.ipynb for ocean heat content calculations and xgcm grid calculation (including integration) https://github.com/xgcm/xgcm/blob/master/xgcm/grid.py) <br>

things to improve: 1. mask land points with nan, 2 round steric height to a few decimal to save space and memory to read

```python
def pbc_dzt(dz,kmt,ht,z_w_bot,mval=0):
    '''readme 2021Apr14: compute partial bottom cell(pbc) for dzt, copied from Fred's email on 2021Mar11'''
    nz = np.shape(dz)[0]
    ny,nx = np.shape(kmt)
    dzt = np.zeros((nz,ny,nx)) + dz.values[:,None,None]
    for iz in range(0,nz):
        bottom = (kmt.values==(iz+1))
        belowbottom = (kmt.values<(iz+1))
        count1 = np.count_nonzero(bottom)
        count2 = np.count_nonzero(belowbottom)
        if (count1 > 0):
            tmp2 = dzt[iz,:,:]
            tmp2[bottom] = ht.values[bottom] - z_w_bot.values[iz-1]
            dzt[iz,:,:]=tmp2
        if (count2 > 0):
            tmp2 = dzt[iz,:,:]
            tmp2[belowbottom] = mval
            dzt[iz,:,:]=tmp2
    dzt = xr.DataArray(dzt,dims=['z_t','nlat','nlon'])
    dzt.encoding['_FillValue']=mval
    return dzt
```

```python
dzt_x = pbc_dzt(ds0.dz,ds0.KMT,ds0.HT,ds0.z_w_bot,mval=0) # x: xarray data array
dzt = dzt_x.chunk(chunks=chunk_size) #convert xarray data array to dask array, there's mem issue otherwise
del dzt_x
dzt
```

# loop to read data and data analysis

```python
%%time 
# ~10min for 2 yrs HR, with 12 cores and 48G memory,
for iyr in yrs: 
    print('working on year',str(iyr))
    infiles = [indir + prefix + str(iyr) + '-' + str(imon).zfill(2) + '.nc' for imon in mons]
    ds = xr.open_mfdataset(infiles, compat="override", combine="by_coords", data_vars="minimal",
                          coords="minimal", chunks=chunk_size, drop_variables=droplist, parallel=True)
    time_s = ds.time.to_pandas().index.shift(-1,'M') # s: shift time
    ds = ds.assign_coords(time=time_s)
    
    print('computing b')
    rho = pop_tools.eos(ds.SALT, ds.TEMP, depth=ds0.z_t*1e-2) # in-situ density rho
    b = 1-rho/rho0 
    
    print('computing steric height')
    h_rst = (b*dzt).sum(dim='z_t').resample(time='A').mean('time') # h_rst: regional steric height
    h_rst = h_rst.where(ds0.KMT>0, np.nan).rename('h_rst') # mask land with nan
    h_rst.attrs['long name'] = 'yr-av regional steric height'
    h_rst.attrs['units'] = 'cm'
    h_rst.attrs['source code'] = 'saveRegionalStericHeightSunway_2021Apr14.ipynb'    
    
    outfile=(outdir + prefix + 'RegionalStericHeight_' + str(iyr) + 'av_' + date_str + '.nc')
    h_rst.to_netcdf(path=outfile, mode='w', format='NETCDF4', compute=True)
    
    del infiles, ds, time_s, rho, b, h_rst, outfile
```

compute difference in z_t to get DZT,  I do not use this way to get DZT for two reasons: 1 CESM POP outputs dzw (midpoint of k to midpoint of k+1), it's most accurate to use model outputs, 2 pop_tools.to_xgcm_grid_dataset() raises error message on dimension when I put DZT into dataset. 

```python
z_t_dif=ds0.z_t.diff('z_t')
DZT=xr.concat([z_t_dif[0], z_t_dif], dim="z_t")
DZT=DZT.assign_coords(z_t=ds0.z_t)
DZT

ds0["DZT"] = DZT
ds0
```

z_t: depth from surface to midpoint of layer (UVEL, VVEL, TEMP, SALT are on z_t, WVEL is on z_w_top) <br>
z_w: depth from surface to top of layer <br>
z_w_top: depth from surface to top of layer <br>
z_w_bot: depth from surface to bottom of layer <br>
dz:  thickness of layer k <br>
dzw: midpoint of k to midpoint of k+1 <br>
