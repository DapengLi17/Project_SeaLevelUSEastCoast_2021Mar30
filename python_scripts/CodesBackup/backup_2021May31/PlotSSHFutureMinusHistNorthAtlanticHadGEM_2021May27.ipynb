```python
mon1 = np.array([1,4,7,10])

yrs1 = np.arange(1991,2010+1)
indir1_TH = '/ihesp/user/dapengli/CMIP6/HadGEM3/HH/hist_1950/'
prefix1_TH = 'zos_Omon_HadGEM3-GC31-HH_hist-1950_r1i1p1f1_gn_'
infiles1_TH = [indir1_TH + prefix1_TH + str(iyr) + str(imon).zfill(2) + '-' + str(iyr) + str(imon+2).zfill(2) +'.nc' 
               for iyr, imon in product(yrs1, mon1)]
# infiles1_TH

yrs2 = np.arange(2031,2050+1)
indir2_TH = '/ihesp/user/dapengli/CMIP6/HadGEM3/HH/future/'
prefix2_TH = 'zos_Omon_HadGEM3-GC31-HH_highres-future_r1i1p1f1_gn_'
infiles2_TH = [indir2_TH + prefix2_TH + str(iyr) + str(imon).zfill(2) + '-' + str(iyr) + str(imon+2).zfill(2) +'.nc' 
               for iyr, imon in product(yrs2, mon1)]

# indir2_TH = '/ihesp/user/dapengli/CMIP6/HadGEM3/HH/future/'
ds1_TH = xr.open_mfdataset(infiles1_TH, compat="override", combine="by_coords", 
                            data_vars="minimal", coords="minimal", parallel=True) # chunks=chunks_HR,
ds2_TH = xr.open_mfdataset(infiles2_TH, compat="override", combine="by_coords", 
                            data_vars="minimal", coords="minimal", parallel=True) # chunks=chunks_HR,
ds2_TH
```

```python
yrs1 = np.arange(1991,2010+1)
indir1_TL = '/ihesp/user/dapengli/CMIP6/HadGEM3/LL/hist_1950/'
prefix1_TL = 'zos_Omon_HadGEM3-GC31-LL_hist-1950_r1i1p1f1_gn_'
infiles1_TL = [indir1_TL + prefix1_TL + str(iyr) + '01-' + str(iyr) + '12.nc' 
               for iyr in yrs1]
# infiles1_TH

yrs2 = np.arange(2031,2050+1)
indir2_TL = '/ihesp/user/dapengli/CMIP6/HadGEM3/LL/future/'
prefix2_TL = 'zos_Omon_HadGEM3-GC31-LL_highres-future_r1i1p1f1_gn_'
infiles2_TL = [indir2_TL + prefix2_TL + str(iyr) + '01-' + str(iyr) + '12.nc' 
               for iyr in yrs2]

# indir2_TH = '/ihesp/user/dapengli/CMIP6/HadGEM3/HH/future/'
ds1_TL = xr.open_mfdataset(infiles1_TL, compat="override", combine="by_coords", 
                            data_vars="minimal", coords="minimal", parallel=True) # chunks=chunks_HR,
ds2_TL = xr.open_mfdataset(infiles2_TL, compat="override", combine="by_coords", 
                            data_vars="minimal", coords="minimal", parallel=True) # chunks=chunks_HR,
ds2_TL
```

```python
ssh1_TH = (ds1_TH.zos.mean('time')*100).compute() # unit: cm
ssh2_TH = (ds2_TH.zos.mean('time')*100).compute() # unit: cm
ssh1_TL = (ds1_TL.zos.mean('time')*100).compute() # unit: cm
ssh2_TL = (ds2_TL.zos.mean('time')*100).compute() # unit: cm

dssh_TH = ssh2_TH - ssh1_TH # unit: cm
dssh_TL = ssh2_TL - ssh1_TL # unit: cm

dssh_TH
```

```python
fig, axs = plt.subplots(1,2,figsize=(15,5))
dssh_TH.plot(ax=axs[0], vmin=-20, vmax=20, cmap='bwr')
dssh_TL.plot(ax=axs[1], vmin=-20, vmax=20, cmap='bwr')
```
