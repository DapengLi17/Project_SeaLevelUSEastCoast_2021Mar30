```python
# convert VVEL(grid y direction) to u2 (north-south,meridional direction)
angle_sec1_TL = ds0_TL.ANGLE.isel(nlat=ilat1_g16,nlon=slice(ilon1_g16,ilon2_g16)).values # [rad]
angle_sec1_TH = ds_vvel_TH.ANGLE.isel(nlat=ilat1_t12,nlon=slice(ilon1_t12,ilon2_t12)).values # [rad]

# v1r_r_TH = np.sin(angle_r_TH)*u1_r_TH + np.cos(angle_r_TH)*v1_r_TH 

u2_sec1_TL = np.sin(angle_sec1_TL)*u_sec1_TL + np.cos(angle_sec1_TL)*v_sec1_TL 
u2_sec1_CL = np.sin(angle_sec1_TL)*u_sec1_CL + np.cos(angle_sec1_TL)*v_sec1_CL
u2_sec1_TH = np.sin(angle_sec1_TH)*u_sec1_TH + np.cos(angle_sec1_TH)*v_sec1_TH 
u2_sec1_CH = np.sin(angle_sec1_TH)*u_sec1_CH + np.cos(angle_sec1_TH)*v_sec1_CH 
```

```python
%%time 
_, nz_sec1_t12, nlon_sec1_t12 = u2_sec1_THmCH.shape
u2_tr_sec1_THmCH = np.empty((nz_sec1_t12, nlon_sec1_t12))
u2_tr_sec1_THmCH[:] = np.nan
p_sec1_THmCH = np.empty((nz_sec1_t12, nlon_sec1_t12))
p_sec1_THmCH[:] = np.nan

for ilon in np.arange(0,nlon_sec1_t12):
    for iz in np.arange(0,nz_sec1_t12):
#         slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
        u2_tr_sec1_THmCH[iz,ilon], _, _, p_sec1_THmCH[iz,ilon], _ = stats.linregress(
            time2, u2_sec1_THmCH[:,iz,ilon])
    
print(u2_tr_sec1_THmCH.shape)
print(u2_sec1_TH.z_t.shape)
print(u2_sec1_TH.z_t.values[:,np.newaxis].shape)
z_sec1_THmCH = np.tile(u2_sec1_TH.z_t.values[:,np.newaxis],(1,len(u2_sec1_TH.ULONG)))
print(z_sec1_THmCH.shape)

print(u2_sec1_TH.ULONG.shape)
print(u2_sec1_TH.ULONG.values[np.newaxis,:].shape)
lon_sec1_THmCH = np.tile(u2_sec1_TH.ULONG.values[np.newaxis,:],(len(u2_sec1_TH.z_t),1))
print(lon_sec1_THmCH.shape)

zm_sec1_THmCH = np.where(p_sec1_THmCH>0.1, z_sec1_THmCH, np.nan) # m: mask
lonm_sec1_THmCH = np.where(p_sec1_THmCH>0.1,lon_sec1_THmCH, np.nan) # m: mask
```
