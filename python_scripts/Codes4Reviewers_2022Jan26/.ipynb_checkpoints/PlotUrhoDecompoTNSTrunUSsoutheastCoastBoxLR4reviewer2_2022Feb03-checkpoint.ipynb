```python
print(ds_hdift_TL.time[0].values, ds_hdift_TL.time[-1].values, len(ds_hdift_TL.time)) 
# print(ds_hdifs_TL.time[0].values, ds_hdifs_TL.time[-1].values, len(ds_hdifs_TL.time)) 
print(ds_uet_TL.time[0].values, ds_uet_TL.time[-1].values, len(ds_uet_TL.time)) 

time_uet_s_TL = ds_uet_TL.time.to_pandas().index.shift(-1,'D') # shift it one month back 
ds_uet_TL = ds_uet_TL.assign_coords(time=time_uet_s_TL)
ds_vnt_TL = ds_vnt_TL.assign_coords(time=time_uet_s_TL)

ds_hdift_TL = ds_hdift_TL.assign_coords(time=time_uet_s_TL)
ds_hdifs_TL = ds_hdifs_TL.assign_coords(time=time_uet_s_TL)

ds_ues_TL = ds_ues_TL.assign_coords(time=time_uet_s_TL)
ds_vns_TL = ds_vns_TL.assign_coords(time=time_uet_s_TL)

print(ds_hdift_TL.time[0].values, ds_hdift_TL.time[-1].values, len(ds_hdift_TL.time)) 
print(ds_uet_TL.time[0].values, ds_uet_TL.time[-1].values, len(ds_uet_TL.time)) 
```

```python
dzt_3d_LR = xr.DataArray(ds_hdift_TL.dz.values[:,None,None]*np.ones((len(ds_hdift_TL.dz), len(ds_hdift_TL.nlat), len(ds_hdift_TL.nlon))), 
                         dims=['z_t','nlat','nlon'], coords={'z_t':ds_hdift_TL.z_t, 'nlat':ds_hdift_TL.nlat, 'nlon':ds_hdift_TL.nlon})

ds_hdift_TL["HDIFS"] = ds_hdifs_TL.HDIFS
ds_hdift_TL["DZU"] = dzt_3d_LR # cm
ds_hdift_TL["DZT"] = dzt_3d_LR # cm
ds_hdift_TL.DZT.attrs["long_name"] = "Thickness of T cells"
ds_hdift_TL.DZT.attrs["units"] = "centimeter"
ds_hdift_TL.DZT.attrs["grid_loc"] = "3111"
ds_hdift_TL.DZU.attrs["long_name"] = "Thickness of U cells"
ds_hdift_TL.DZU.attrs["units"] = "centimeter"
ds_hdift_TL.DZU.attrs["grid_loc"] = "3221"

ds_uet_TL["VNT"] = ds_vnt_TL.VNT
ds_uet_TL["UES"] = ds_ues_TL.UES
ds_uet_TL["VNS"] = ds_vns_TL.VNS
ds_uet_TL["DZU"] = dzt_3d_LR # cm
ds_uet_TL["DZT"] = dzt_3d_LR # cm
ds_uet_TL.DZT.attrs["long_name"] = "Thickness of T cells"
ds_uet_TL.DZT.attrs["units"] = "centimeter"
ds_uet_TL.DZT.attrs["grid_loc"] = "3111"
ds_uet_TL.DZU.attrs["long_name"] = "Thickness of U cells"
ds_uet_TL.DZU.attrs["units"] = "centimeter"
ds_uet_TL.DZU.attrs["grid_loc"] = "3221"

# make sure we have the cell volumne for calculations
VOL = (ds_uet_TL.DZT * ds_uet_TL.DXT * ds_uet_TL.DYT).compute()
KMT = ds_uet_TL.KMT.compute()

for j in tqdm(range(len(KMT.nlat))):
    for i in range(len(KMT.nlon)):
        k = KMT.values[j, i].astype(int)
        VOL.values[k:, j, i] = 0.0

ds_uet_TL["VOL"] = VOL

ds_uet_TL.VOL.attrs["long_name"] = "volume of T cells"
ds_uet_TL.VOL.attrs["units"] = "centimeter^3"

ds_uet_TL.VOL.attrs["grid_loc"] = "3111"
```

```python
%%time

nlon1_r_g16, nlon2_r_g16 = 260+20, 321-30
nlat1_r_g16, nlat2_r_g16 = 230+35, 350-50

metrics = {
    ("X",): ["DXU", "DXT"],  # X distances
    ("Y",): ["DYU", "DYT"],  # Y distances
    ("Z",): ["DZU", "DZT"],  # Z distances
    ("X", "Y"): ["UAREA", "TAREA"],}

ds_hdift_r_TL = ds_hdift_TL.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16), nlat=slice(nlat1_r_g16, nlat2_r_g16))
ds_uet_r_TL = ds_uet_TL.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16), nlat=slice(nlat1_r_g16, nlat2_r_g16))

# ds_hdift_tre_r_TL = ds_hdift_TL.resample(time='A').mean('time')
# ds_uet_tre_r_TL = ds_uet_TL.resample(time='A').mean('time')
ds_uet_r_TL

# xgcm metrics
# gridm, xdsm = pop_tools.to_xgcm_grid_dataset(ds, periodic=False, metrics=metrics)
grid_hdift_r_TL, xds_hdift_r_TL = pop_tools.to_xgcm_grid_dataset(ds_hdift_r_TL, 
                                                                 periodic=False,
                                                                 metrics=metrics,
                                                                 boundary={"X": "extend", "Y": "extend", "Z": "extend"},)


for coord in ["nlat", "nlon"]:
    if coord in xds_hdift_r_TL.coords:
        xds_hdift_r_TL = xds_hdift_r_TL.drop_vars(coord)


grid_uet_r_TL, xds_uet_r_TL = pop_tools.to_xgcm_grid_dataset(ds_uet_r_TL, 
                                                             periodic=False,
                                                             metrics=metrics,
                                                             boundary={"X": "extend", "Y": "extend", "Z": "extend"},)


for coord in ["nlat", "nlon"]:
    if coord in xds_uet_r_TL.coords:
        xds_uet_r_TL = xds_uet_r_TL.drop_vars(coord)
        
xds_uet_r_TL
```
