```python
pc = cartopy.crs.PlateCarree()
lon_formatter = LongitudeFormatter(zero_direction_label=True)
lat_formatter = LatitudeFormatter()

fig1, axs = plot.subplots(nrows=1,ncols=2,proj='cyl',width='8in',height='3.5in')

temp_TH = xds_r_TH.TEMP.isel(time=0,z_t=0).where(xds_r_TH.KMT>0, np.nan)
temp_TL = xds_r_TL.TEMP.isel(time=0,z_t=0).where(xds_r_TL.KMT>0, np.nan)

m00=axs[0,0].pcolormesh(xds_r_g16.TLONG, xds_r_g16.TLAT, 
        temp_TL,levels=np.arange(0, 25+0.1, 5), 
        transform=pc, cmap="ColdHot", extend='both')
axs[0,0].plot(xds_r_g16.TLONG.isel(nlat_t=41, nlon_t=slice(23,26)), xds_r_g16.TLAT.isel(nlat_t=41, nlon_t=slice(23,26)),  
        transform=pc, color='k')
axs[0,0].plot(xds_r_g16.TLONG.isel(nlat_t=slice(41,60), nlon_t=25), xds_r_g16.TLAT.isel(nlat_t=slice(41,60), nlon_t=25),  
        transform=pc, color='k')
axs[0,0].coastlines(resolution='110m', color='black', linewidth=1)
axs[0,0].format(land=True, landcolor='grey')
axs[0,0].set_extent([lon1_r, lon2_r, lat1_r, lat2_r], crs=pc)
axs[0,0].set_xticks(np.arange(lon1_r,lon2_r+1,10), crs=pc)
axs[0,0].set_xticklabels(np.arange(lon1_r,lon2_r+1,10), fontsize=12)
axs[0,0].xaxis.set_major_formatter(lon_formatter)
axs[0,0].set_yticks(np.arange(lat1_r,lat2_r+1, 10), crs=pc)
axs[0,0].set_yticklabels(np.arange(lat1_r,lat2_r+1, 10), fontsize=12)
axs[0,0].yaxis.set_major_formatter(lat_formatter)
axs[0,0].set_title('TEMP LR', fontsize=12, fontweight='bold')
axs[0,0].text(-84, 53, 'a', transform=pc, fontsize=12)


m01=axs[0,1].pcolormesh(xds_r_t12.TLONG, xds_r_t12.TLAT, 
        temp_TH,levels=np.arange(0, 25+0.1, 5), 
        transform=pc, cmap="ColdHot", extend='both')
axs[0,1].plot(xds_r_t12.TLONG.isel(nlat_t=58, nlon_t=slice(45,66)), xds_r_t12.TLAT.isel(nlat_t=58, nlon_t=slice(45,66)),  
        transform=pc,color='k')
axs[0,1].plot(xds_r_t12.TLONG.isel(nlat_t=slice(58,150), nlon_t=65), xds_r_t12.TLAT.isel(nlat_t=slice(58,150), nlon_t=65),  
        transform=pc,color='k')
axs[0,1].coastlines(resolution='110m', color='black', linewidth=1)
axs[0,1].format(land=True, landcolor='grey')
axs[0,1].set_extent([lon1_r, lon2_r, lat1_r, lat2_r], crs=pc)
axs[0,1].set_xticks(np.arange(lon1_r,lon2_r+1,10), crs=pc)
axs[0,1].set_xticklabels(np.arange(lon1_r,lon2_r+1,10), fontsize=12)
axs[0,1].xaxis.set_major_formatter(lon_formatter)
axs[0,1].set_yticks(np.arange(lat1_r,lat2_r+1, 10), crs=pc)
axs[0,1].set_yticklabels(np.arange(lat1_r,lat2_r+1, 10), fontsize=12)
axs[0,1].yaxis.set_major_formatter(lat_formatter)
axs[0,1].set_title('TEMP HR', fontsize=12, fontweight='bold')
axs[0,1].text(-84, 53, 'd', transform=pc, fontsize=12)
# add colorbar
cb = axs[0,1].colorbar(m01,loc='r', label='[degC]',width='2em')
```
