```python
ds0 = xr.open_dataset(infile0, chunks=chunk_size) # reference file 
ds0
```

```python
droplist = list(ds0.variables)
keeplist=['UVEL','VVEL','TEMP','SALT','time','TLONG','TLAT','ULONG','ULAT','z_t','z_w',
          'KMT','KMU','dz','dzw','DXU','DYU','TAREA','UAREA']
for i in keeplist:
    droplist.remove(i)
droplist
```

```python
KMT_NA = ds0.KMT.isel(nlon=nlon_NA_slice, nlat=nlat_NA_slice)
KMU_NA = ds0.KMU.isel(nlon=nlon_NA_slice, nlat=nlat_NA_slice)

z_t_dif = ds0.z_t.diff('z_t').values
print(z_t_dif)
z_t_dif = np.insert(z_t_dif, 0, z_t_dif[0])
DZT = xr.DataArray(data=z_t_dif, coords=[ds0.z_t.values], dims=["z_t"], attrs=dict(longname="T-grid cell thickness",units="cm"))
DZT = DZT.rename('DZT')
DZT
```

```python
# using xgcm
def div_4pt_xgcm(U, V):  # Basic stencil operator for B-grid divergence
    """
    POP stencil operator for divergence
    using xgcm
    """
    U_at_lat_t = grid.interp(U, axis="Y", boundary="extend")  # array - array.roll becomes interp
    # -> this incoorporates the * 0.5
    dUdx = grid.diff(U_at_lat_t, axis="X", boundary="extend")  # - becomes diff
    V_at_lon_t = grid.interp(V, axis="X", boundary="extend")
    dVdy = grid.diff(V_at_lon_t, axis="Y", boundary="extend")
    div_4pt = dUdx + dVdy
    return div_4pt


def divh_xgcm_DL(U, V, ds_grid):  # Divergence of horizontal velocity. Result on T-grid
    """
    calulate divergence with xgcm
    2021Mar08 edited by DL: remove ds_grid.DZU and ds_grid.DZT, Qiuying said DZT and DZU are the same
    """
    changed_U = U * ds_grid.DYU
    changed_V = V * ds_grid.DXU
    divh = div_4pt_xgcm(changed_U, changed_V) / ds_grid.TAREA
    return divh
```

```python
%%time
# 14min for 12 cores to analyze 10 yr data files
for iyr in yrs:
    print('working on year ',str(iyr))
    infiles = [indir + prefix + str(iyr) + '-' + str(imon).zfill(2) + '.nc' 
              for imon in mons]
    ds = xr.open_mfdataset(infiles, compat="override", combine="by_coords",data_vars="minimal",
                    coords="minimal", chunks=chunk_size, drop_variables=droplist, parallel=True)
    ds_NA = ds.isel(nlon=nlon_NA_slice, nlat=nlat_NA_slice)

    grid, ds_NA_x = pop_tools.to_xgcm_grid_dataset(ds_NA, periodic=False)
    
    u1_NA = ds_NA_x.UVEL.where(ds_NA_x.KMU>0, np.nan)  # mask land as nan
    u2_NA = ds_NA_x.VVEL.where(ds_NA_x.KMU>0, np.nan)  # mask land as nan
    salt_NA = ds_NA_x.SALT.where(ds_NA_x.KMT>0, np.nan)  # mask land as nan
    temp_NA = ds_NA_x.TEMP.where(ds_NA_x.KMT>0, np.nan)  # mask land as nan
    
    rho_NA = pop_tools.eos(salt_NA, temp_NA, depth=ds_NA_x.z_t * 1e-2)
    rho_NA_u = grid.interp(rho_NA, ["X","Y"], boundary="extend").rename('density_U')
    
    mf1_NA = (rho_NA_u*u1_NA*DZT).sum(dim='z_t', skipna=True) # mass (m) flux (f)
    mf2_NA = (rho_NA_u*u2_NA*DZT).sum(dim='z_t', skipna=True) # mass (m) flux (f)
    
    mf_div_NA = divh_xgcm_DL(mf1_NA, mf2_NA, ds_NA_x) # mass (m) flux (f)
    mt_vel_NA = (-mf_div_NA/rho0).rename('TEND_MASS_VEL') # mass (m) tendency (t)
    mt_vel_NA.attrs['unit'] = 'cm/s'
    mt_vel_NA.attrs['long name'] = 'mass tendency due to flow convergence, see Griffies et al., 2014 Eq (48-49)'
    mt_vel_NA.attrs['source code'] = 'saveMassTendConvNorthAtlantic1950_2050YearAv_HR_LR_2021Mar08.ipynb'

    outfile = outdir + prefix + '_MassTendConvNorthAtlantic_' + str(iyr) + '_' + date_str + '.nc'

    mt_vel_NA.to_netcdf(path=outfile, mode='w', format='NETCDF4', compute=True) 
        
    del infiles, ds, ds_NA, grid, ds_NA_x, u1_NA, u2_NA, temp_NA, salt_NA, rho_NA, rho_NA_u
    del mf1_NA, mf2_NA, mf_div_NA, mt_vel_NA, outfile
```
