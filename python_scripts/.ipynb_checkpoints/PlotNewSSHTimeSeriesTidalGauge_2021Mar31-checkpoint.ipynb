```python
infile0_HR = ('/scratch/group/ihesp/archive/B.E.13.BHISTC5.ne120_t12.sehires38.003.sunway/ocn/monthly/'
              'B.E.13.BHISTC5.ne120_t12.sehires38.003.sunway.pop.h.1850-01.nc')
ds0_HR = xr.open_dataset(infile0_HR,chunks={'nlat':600,'nlon':600})
ds0_HR
```

```python
ds0_HR.sel(nlat=loc1_HR-3, nlon=loc2_HR).KMT.values
```

```python
ds0_HR.KMT.plot()
ds0_HR.sel(nlat=loc1_HR-2, nlon=loc2_HR).KMT.values
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,4))
ds0_HR.TLONG.plot(ax=axes[0])
ds0_HR.TLAT.plot(ax=axes[1])
```

```python
# A 2D plot of the SPEED variable, assigning the coordinate values,
# and plot the verticies of each point
# ds0_LR.KMT.plot()
plt.scatter(ds0_HR.TLONG, ds0_HR.TLAT)

# I want to find the speed at a certain lat/lon point.
lat,lon = 40.7, -74.0+360 # lat and lon of New York

# First, find the index of the grid point nearest a specific lat/lon.   
abslat = np.abs(ds0_HR.TLAT-lat)
abslon = np.abs(ds0_HR.TLONG-lon)
c = np.maximum(abslon, abslat)

([loc1_HR], [loc2_HR]) = np.where(c == np.min(c)) 
#DL loc1_HR, loc2_HR are the index for the 1st and 2nd dimensions
#DL 1st dimension is nlat and 2nd dimension is nlon

# Now I can use that index location to get the values at the x/y diminsion
point_ds = ds0_HR.sel(nlat=loc1_HR, nlon=loc2_HR)

# Plot requested lat/lon point blue
plt.scatter(lon, lat, color='b')
plt.text(lon, lat, 'requested')

# Plot nearest point in the array red
plt.scatter(point_ds.TLONG, point_ds.TLAT, color='r')
plt.text(point_ds.TLONG, point_ds.TLAT, 'nearest')

# plt.title('speed at nearest point: %s' % point_ds.SPEED.data)
print('index for the first and second dimension')
print(loc1_HR,loc2_HR)
```

```python
infile_t12 = '/ihesp/user/dapengli/Projects4iHESP/SupplementaryFiles4POP/POPgrids_tx0.1v2_2020Jun29.nc'
ds_t12 = xr.open_dataset(infile_t12)
ds_t12
```

```python
# output grid, 1x1 degree rectilinear
ds_out = xe.util.grid_global(1, 1)
ds_out
```

```python
# --- regrid HR ---
ds_t12 = ds_t12.rename({'TLONG': 'lon', 'TLAT': 'lat'})

weight_file_HR=('/ihesp/user/dapengli/Projects4iHESP/SupplementaryFiles4POP/'
             'bilinear_t12_1x1degGlobalRectlinear_periodic_2020Aug01.nc')
regridder_HR = xe.Regridder(ds_t12, ds_out, method='bilinear', periodic=True, filename=weight_file_HR, reuse_weights=True)
# reuse weight file. 
regridder_HR
```

```python
ssh_tmc1_HR = ds_tr1Av_HR.SSH.isel(time=0) - ds_c1Av_HR.SSH.isel(time=0) # tmc: transient minus control
ssh_tmc2_HR = ds_tr2Av_HR.SSH.isel(time=0) - ds_c2Av_HR.SSH.isel(time=0)
ssh_tmc1_rgd_HR = regridder_HR(ssh_tmc1_HR) # rgd: regrid
ssh_tmc2_rgd_HR = regridder_HR(ssh_tmc2_HR) # rgd: regrid
ssh_tmc2_rgd_HR
```

```python
# --- regrid LR ---
ds_g16 = ds_g16.rename({'TLONG': 'lon', 'TLAT': 'lat'})

# 
weight_file_LR=('/ihesp/user/dapengli/Projects4iHESP/SupplementaryFiles4POP/'
             'bilinear_g16_1x1degGlobalRectlinear_periodic_2020Aug01.nc')
regridder_LR = xe.Regridder(ds_g16, ds_out, method='bilinear', periodic=True, filename=weight_file_LR, reuse_weights=True)
# reuse weight file. 
regridder_LR
```

```python
ssh_tmc1_LR = ds_tr_LR.SSH.sel(time=slice('1951','1970')).mean('time') - ds_c_LR.SSH.sel(time=slice('1951','1970')).mean('time')
ssh_tmc2_LR = ds_tr_LR.SSH.sel(time=slice('2031','2050')).mean('time') - ds_c_LR.SSH.sel(time=slice('2031','2050')).mean('time')
ssh_tmc1_rgd_LR = regridder_LR(ssh_tmc1_LR) # rgd: regrid
ssh_tmc2_rgd_LR = regridder_LR(ssh_tmc2_LR) # rgd: regrid
ssh_tmc2_rgd_LR
```
