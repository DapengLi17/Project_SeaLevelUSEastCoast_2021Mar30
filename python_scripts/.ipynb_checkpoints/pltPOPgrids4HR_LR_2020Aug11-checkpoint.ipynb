# import modules

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

# plot HR CESM-POP

```python
infile_HR = ('/ihesp/archive/HRMIP/B.E.13.B1950TRC5.ne120_t12.cesm-ihesp-1950-2050.013/ocn/hist/'
             'B.E.13.B1950TRC5.ne120_t12.cesm-ihesp-1950-2050.013.pop.h.nday1.2050-09-01.nc')
ds_HR = xr.open_dataset(infile_HR)
print(ds_HR)
print('=========================')
# CESM POP uses -1 for lat of land, I use lat and lon from an extra grid file to plot (see below)
print(ds_HR.TLAT.values) 
print('===========================')
# CESM POP uses -1 for lon of land, I use lat and lon from an extra grid file to plot (see below)
print(ds_HR.TLONG.values) 
print('===========================')
```

```python
sst_HR_raw = ds_HR.SST.isel(time=0)
print(sst_HR_raw) # TLAT and TLONG are also -1 for land
print('===========================')
```

```python
# === input grids ===
POPgridsFile_tx01v2 = '/ihesp/user/dapengli/Projects4iHESP/Project_MLD_2020May28/DataAnalysis/raw_data/POPgrids/POPgrids_tx0.1v2_2020Jun29.nc'
ds_grid_tx01v2 = xr.open_dataset(POPgridsFile_tx01v2)
print(ds_grid_tx01v2)
print('========================')
print(ds_grid_tx01v2.TLAT.values) # real lat and lon for lands 
print('========================')
print(ds_grid_tx01v2.TLONG.values) # real lat and lon for lands 
print('========================')
# ====================
```

```python
print(sst_HR_raw)
print('=============')
print(ds_HR.KMT)
print('=============')
sst_HR_raw2 = sst_HR_raw.where(ds_HR.KMT>0,np.nan) # use ds_HR.KMT to mask land with NaN
print(sst_HR_raw2)
print('========================')
```

```python
proj = ccrs.Robinson(central_longitude=0)
pc = ccrs.PlateCarree()

fig = plt.figure(figsize=(15, 8))

ax1 = fig.add_subplot(121,projection=proj)
subplt1=ax1.pcolormesh(ds_grid_tx01v2.TLONG, ds_grid_tx01v2.TLAT, sst_HR_raw2,
                       vmin=-10,vmax=40,transform=pc,cmap='jet')
```

# plot LR CESM-POP

```python
infile_LR = ('/ihesp/archive/LR/b.e13.B1950C5.ne30_g16.ihesp24_1950cntl.002/ocn/hist/'
             'b.e13.B1950C5.ne30_g16.ihesp24_1950cntl.002.pop.h.nday1.0066-07-01.nc')
ds_LR = xr.open_dataset(infile_LR)
print(ds_LR)
print('==========================')
# CESM LR POP uses real lat and lon for land, so no need to use an extra grid file to plot
print(ds_LR.TLAT.values) 
print('===========================')
print(ds_LR.TLONG.values) 
print('===========================')
```

```python
sst_LR_raw = ds_LR.SST.isel(time=0) # variable to be regrided
print(sst_LR_raw)
print('===========================')
sst_LR_raw2 = sst_LR_raw.where(ds_LR.KMT>0,np.nan) # mask land with NaN
print(sst_LR_raw2)
print('===========================')
```

```python
proj = ccrs.Robinson(central_longitude=0)
pc = ccrs.PlateCarree()

fig = plt.figure(figsize=(15, 8))

ax1 = fig.add_subplot(121,projection=proj)
subplt1=ax1.pcolormesh(ds_LR.TLONG, ds_LR.TLAT, sst_LR_raw2,
                       vmin=-10,vmax=40,transform=pc,cmap='jet')
```
