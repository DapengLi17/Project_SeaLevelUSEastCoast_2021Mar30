```python
infile_ssh_obs = ('/scratch/user/dapengli/Projects4iHESP/Project_SeaLevelUSEastCoast_2021Mar30/data_after_manipulation/'
                  'ADT_Vgeo_CMEMS/dt_global_allsat_phy_l4_1993-2019_2021May18.nc')

ds_ssh_raw_obs = xr.open_dataset(infile_ssh_obs)
jultime_ssh_obs = ds_ssh_raw_obs.time.to_pandas().index.to_julian_date() # convert datetime to julian date 
ds_ssh_obs = ds_ssh_raw_obs.assign_coords(time=jultime_ssh_obs)
ds_ssh_obs
```

```python
weight_ssh_obs = np.cos(np.deg2rad(ds_ssh_obs.latitude))
weight_ssh_obs.name = "weight_ssh_obs"
ssh_gl_av_obs = ds_ssh_obs.adt.weighted(weight_ssh_obs).mean(("longitude", "latitude")) # unit: m
ssh_gl_av_obs.plot()
ssh_gl_av_obs
```

```python
# print(ds_ssh_obs.time)
# jultime_ssh_obs = ds_ssh_obs.time.to_pandas().index.to_julian_date() # convert datetime to julian date 
# print(jultime_ssh_obs)
# ssh_gl_av_obs = ssh_gl_av_obs.assign_coords(time=jultime_ssh_obs)
# print(ssh_gl_av_obs.time)# unit: m/day to mm/yr
ssh_gl_av_tr_obs = ssh_gl_av_obs.polyfit(dim='time',deg=1).polyfit_coefficients.sel(degree=1)*1000*365
ssh_gl_av_tr_obs
```

```python
trend_ssh = ds_ssh_obs.adt.sel(latitude=slice(10,60),
                    longitude=slice(-100+360, -40+360)
                   ).polyfit(dim='time',deg=1).polyfit_coefficients.sel(degree=1)*1000*365 # m/day to mm/yr
# trend = (trend*1000*365).compute() 
trend_ssh = trend_ssh.rename('trend_ssh')
trend_ssh.attrs['LongName'] = ('linear trend in Altimeter SSH from 1993-2019')
trend_ssh.attrs['units'] = 'mm/yr'
trend_ssh.attrs['SourceCode'] = 'saveSLAtrendsObs_2021Jan18.ipynb'
trend_ssh
(trend_ssh-ssh_gl_av_tr_obs).plot(vmin=-5,vmax=5,cmap='bwr')
```
