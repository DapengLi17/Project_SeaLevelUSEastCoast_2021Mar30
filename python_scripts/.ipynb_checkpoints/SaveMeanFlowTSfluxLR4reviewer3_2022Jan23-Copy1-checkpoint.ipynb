# data analysis

```python
# make TH time coordinate consistent across different ds, otherwise xarray data array subtraction does not work 
print(ds_hdift_TL.time[0].values, ds_hdift_TL.time[-1].values, len(ds_hdift_TL.time)) 
print(ds_hdifs_TL.time[0].values, ds_hdifs_TL.time[-1].values, len(ds_hdifs_TL.time)) 
print(ds_uet_TL.time[0].values, ds_uet_TL.time[-1].values, len(ds_uet_TL.time)) 
print(ds_vnt_TL.time[0].values, ds_vnt_TL.time[-1].values, len(ds_vnt_TL.time)) 
print(ds_ues_TL.time[0].values, ds_ues_TL.time[-1].values, len(ds_ues_TL.time)) 
print(ds_vns_TL.time[0].values, ds_vns_TL.time[-1].values, len(ds_vns_TL.time)) 

ds_hdift_TL = ds_hdift_TL.assign_coords(time=ds_uet_TL.time)
ds_hdifs_TL = ds_hdifs_TL.assign_coords(time=ds_uet_TL.time)
ds_vnt_TL = ds_vnt_TL.assign_coords(time=ds_uet_TL.time)
ds_ues_TL = ds_ues_TL.assign_coords(time=ds_uet_TL.time)
ds_vns_TL = ds_vns_TL.assign_coords(time=ds_uet_TL.time)
# ds_uvel_TH = ds_uvel_TH.assign_coords(time=time_vvel_s_TH)
# ds_vvel_TH = ds_vvel_TH.assign_coords(time=time_vvel_s_TH)
```

```python
dzt_3d_LR = xr.DataArray(ds_hdift_TL.dz.values[:,None,None]*np.ones((len(ds_hdift_TL.dz), len(ds_hdift_TL.nlat), len(ds_hdift_TL.nlon))), 
                         dims=['z_t','nlat','nlon'], coords={'z_t':ds_hdift_TL.z_t, 'nlat':ds_hdift_TL.nlat, 'nlon':ds_hdift_TL.nlon})

ds_hdift_TL["HDIFS"] = ds_hdifs_TL.HDIFS
ds_hdift_TL["DZU"] = dzt_3d_LR # cm
ds_hdift_TL["DZT"] = dzt_3d_LR # cm
ds_hdift_TL.DZT.attrs["long_name"] = "Thickness of T cells"
ds_hdift_TL.DZT.attrs["units"] = "centimeter"
ds_hdift_TL.DZT.attrs["grid_loc"] = "3111"
ds_hdift_TL.DZU.attrs["long_name"] = "Thickness of U cells"
ds_hdift_TL.DZU.attrs["units"] = "centimeter"
ds_hdift_TL.DZU.attrs["grid_loc"] = "3221"

ds_uet_TL["VNT"] = ds_vnt_TL.VNT
ds_uet_TL["UES"] = ds_ues_TL.UES
ds_uet_TL["VNS"] = ds_vns_TL.VNS
ds_uet_TL["DZU"] = dzt_3d_LR # cm
ds_uet_TL["DZT"] = dzt_3d_LR # cm
ds_uet_TL.DZT.attrs["long_name"] = "Thickness of T cells"
ds_uet_TL.DZT.attrs["units"] = "centimeter"
ds_uet_TL.DZT.attrs["grid_loc"] = "3111"
ds_uet_TL.DZU.attrs["long_name"] = "Thickness of U cells"
ds_uet_TL.DZU.attrs["units"] = "centimeter"
ds_uet_TL.DZU.attrs["grid_loc"] = "3221"

# make sure we have the cell volumne for calculations
VOL = (ds_uet_TL.DZT * ds_uet_TL.DXT * ds_uet_TL.DYT).compute()
KMT = ds_uet_TL.KMT.compute()

for j in tqdm(range(len(KMT.nlat))):
    for i in range(len(KMT.nlon)):
        k = KMT.values[j, i].astype(int)
        VOL.values[k:, j, i] = 0.0

ds_uet_TL["VOL"] = VOL

ds_uet_TL.VOL.attrs["long_name"] = "volume of T cells"
ds_uet_TL.VOL.attrs["units"] = "centimeter^3"

ds_uet_TL.VOL.attrs["grid_loc"] = "3111"
```

```python
%%time
metrics = {
    ("X",): ["DXU", "DXT"],  # X distances
    ("Y",): ["DYU", "DYT"],  # Y distances
    ("Z",): ["DZU", "DZT"],  # Z distances
    ("X", "Y"): ["UAREA", "TAREA"],}

ds_hdift_r_TL = ds_hdift_TL.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16), nlat=slice(nlat1_r_g16, nlat2_r_g16))
ds_uet_r_TL = ds_uet_TL.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16), nlat=slice(nlat1_r_g16, nlat2_r_g16))

# ds_hdift_tre_r_TL = ds_hdift_TL.resample(time='A').mean('time')
# ds_uet_tre_r_TL = ds_uet_TL.resample(time='A').mean('time')
ds_uet_r_TL

# xgcm metrics
# gridm, xdsm = pop_tools.to_xgcm_grid_dataset(ds, periodic=False, metrics=metrics)
grid_hdift_r_TL, xds_hdift_r_TL = pop_tools.to_xgcm_grid_dataset(ds_hdift_r_TL, 
                                                                 periodic=False,
                                                                 metrics=metrics,
                                                                 boundary={"X": "extend", "Y": "extend", "Z": "extend"},)


for coord in ["nlat", "nlon"]:
    if coord in xds_hdift_r_TL.coords:
        xds_hdift_r_TL = xds_hdift_r_TL.drop_vars(coord)


grid_uet_r_TL, xds_uet_r_TL = pop_tools.to_xgcm_grid_dataset(ds_uet_r_TL, 
                                                             periodic=False,
                                                             metrics=metrics,
                                                             boundary={"X": "extend", "Y": "extend", "Z": "extend"},)


for coord in ["nlat", "nlon"]:
    if coord in xds_uet_r_TL.coords:
        xds_uet_r_TL = xds_uet_r_TL.drop_vars(coord)
        
xds_uet_r_TL
```

```python
# xds_uet_r_TL.UET # [degC/s]
print(xds_uet_r_TL.UET.dims) # 'nlat_t', 'nlon_u'
print(xds_uet_r_TL.UET.resample(time='A').mean('time').dims) # 'nlat_t', 'nlon_u'
grid_uet_r_TL.diff(xds_uet_r_TL.UET, axis="X").dims
# uet_tre_r_TL.dims
# xds_uet_r_TL.VOL.values
# xds_uet_r_TL.VOL
```

```python
uet_tre_r_TL = xds_uet_r_TL.UET.resample(time='A').mean('time')
vnt_tre_r_TL = xds_uet_r_TL.VNT.resample(time='A').mean('time')
ues_tre_r_TL = xds_uet_r_TL.UES.resample(time='A').mean('time')
vns_tre_r_TL = xds_uet_r_TL.VNS.resample(time='A').mean('time')
uet_tre_r_TL
```

```python
duet_tre_r_TL = -(grid_uet_r_TL.diff(uet_tre_r_TL * xds_uet_r_TL.VOL.values, axis="X") / xds_uet_r_TL.VOL)
dvnt_tre_r_TL = -(grid_uet_r_TL.diff(vnt_tre_r_TL * xds_uet_r_TL.VOL.values, axis="Y") / xds_uet_r_TL.VOL)
dues_tre_r_TL = -(grid_uet_r_TL.diff(ues_tre_r_TL * xds_uet_r_TL.VOL.values, axis="X") / xds_uet_r_TL.VOL)
dvns_tre_r_TL = -(grid_uet_r_TL.diff(vns_tre_r_TL * xds_uet_r_TL.VOL.values, axis="Y") / xds_uet_r_TL.VOL)
hdivt_tre_r_TL = (duet_tre_r_TL + dvnt_tre_r_TL)
hdivs_tre_r_TL = (dues_tre_r_TL + dvns_tre_r_TL)
hdivt_tre_r_TL 
```

```python
xds_hdift_r_TL.HDIFT.dims
```

```python
hdift_tre_r_TL = xds_hdift_r_TL.HDIFT.resample(time='A').mean('time')
hdifs_tre_r_TL = xds_hdift_r_TL.HDIFS.resample(time='A').mean('time')
hdift_tre_r_TL

hdivt_tre_intz_r_TL = (hdivt_tre_r_TL*xds_uet_r_TL.DZT).sum('z_t') # [degC/s*cm]
hdivs_tre_intz_r_TL = (hdivs_tre_r_TL*xds_uet_r_TL.DZT).sum('z_t') # [psu/s*cm]
hdivt_tre_intz_r_TL
```

```python
fig, axs = plt.subplots(2,3,figsize=(10,6))
hdivt_tre_intz_r_TL.isel(time=0).plot(ax=axs[0,0], robust=True)
hdift_tre_r_TL.isel(time=0).plot(ax=axs[0,1], robust=True)
(hdivt_tre_intz_r_TL.isel(time=0) + hdift_tre_r_TL.isel(time=0)).plot(ax=axs[0,2], robust=True)

hdivs_tre_intz_r_TL.isel(time=0).plot(ax=axs[1,0], robust=True)
hdifs_tre_r_TL.isel(time=0).plot(ax=axs[1,1], robust=True)
(hdivs_tre_intz_r_TL.isel(time=0) + hdifs_tre_r_TL.isel(time=0)).plot(ax=axs[1,2], robust=True)
```
