```python
# output grid, 1x1 degree rectilinear
ds_out = xe.util.grid_global(1, 1)
ds_out
```

```python
# --- regrid HR ---
ds_t12 = ds_t12.rename({'TLONG': 'lon', 'TLAT': 'lat'})

weight_file_HR=('/ihesp/user/dapengli/Projects4iHESP/SupplementaryFiles4POP/'
             'bilinear_t12_1x1degGlobalRectlinear_periodic_2020Aug01.nc')
regridder_HR = xe.Regridder(ds_t12, ds_out, method='bilinear', periodic=True, filename=weight_file_HR, reuse_weights=True)
# reuse weight file. 
regridder_HR
```

```python
ssh_tmc1_HR = ds_tr1Av_HR.SSH.isel(time=0) - ds_c1Av_HR.SSH.isel(time=0) # tmc: transient minus control
ssh_tmc2_HR = ds_tr2Av_HR.SSH.isel(time=0) - ds_c2Av_HR.SSH.isel(time=0)
ssh_tmc1_rgd_HR = regridder_HR(ssh_tmc1_HR) # rgd: regrid
ssh_tmc2_rgd_HR = regridder_HR(ssh_tmc2_HR) # rgd: regrid
ssh_tmc2_rgd_HR
```

```python
# --- regrid LR ---
ds_g16 = ds_g16.rename({'TLONG': 'lon', 'TLAT': 'lat'})

# 
weight_file_LR=('/ihesp/user/dapengli/Projects4iHESP/SupplementaryFiles4POP/'
             'bilinear_g16_1x1degGlobalRectlinear_periodic_2020Aug01.nc')
regridder_LR = xe.Regridder(ds_g16, ds_out, method='bilinear', periodic=True, filename=weight_file_LR, reuse_weights=True)
# reuse weight file. 
regridder_LR
```

```python
ssh_tmc1_LR = ds_tr_LR.SSH.sel(time=slice('1951','1970')).mean('time') - ds_c_LR.SSH.sel(time=slice('1951','1970')).mean('time')
ssh_tmc2_LR = ds_tr_LR.SSH.sel(time=slice('2031','2050')).mean('time') - ds_c_LR.SSH.sel(time=slice('2031','2050')).mean('time')
ssh_tmc1_rgd_LR = regridder_LR(ssh_tmc1_LR) # rgd: regrid
ssh_tmc2_rgd_LR = regridder_LR(ssh_tmc2_LR) # rgd: regrid
ssh_tmc2_rgd_LR
```

# make pics

```python
pc = ccrs.PlateCarree()
axes_class = (GeoAxes, dict(map_projection=pc))

fig = plt.figure(figsize=(16, 8))
axgr = AxesGrid(fig, 111, axes_class=axes_class, nrows_ncols=(2, 2), axes_pad=0.5, cbar_location='right',
               cbar_mode='single', cbar_pad=0.2, cbar_size='3%', label_mode='')  # note the empty label_mode

# p=axgr[0].pcolormesh(ds_g16.TLONG, ds_g16.TLAT, ssh_tmc1_LR,
#                        vmin=-30,vmax=30,transform=pc,cmap="RdBu_r")
axgr[0].pcolormesh(ssh_tmc1_rgd_LR.lon, ssh_tmc1_rgd_LR.lat, ssh_tmc1_rgd_LR,
                       vmin=-30,vmax=30,transform=pc,cmap="RdBu_r")
axgr[0].coastlines()
axgr[0].set_xticks(np.linspace(-180, 180, 5), crs=pc)
axgr[0].set_yticks(np.linspace(-90, 90, 5), crs=pc)
#         lon_formatter = LongitudeFormatter(zero_direction_label=True)
#         lat_formatter = LatitudeFormatter()
#         ax.xaxis.set_major_formatter(lon_formatter)
#         ax.yaxis.set_major_formatter(lat_formatter)
axgr[0].set_title('transient - control (LR 1951-1970 av)')
axgr[0].tick_params(labelbottom=False)

# axgr[1].pcolormesh(ds_g16.TLONG, ds_g16.TLAT, ssh_tmc2_LR,
#                        vmin=-30,vmax=30,transform=pc,cmap="RdBu_r")
axgr[1].pcolormesh(ssh_tmc2_rgd_LR.lon, ssh_tmc2_rgd_LR.lat, ssh_tmc2_rgd_LR,
                       vmin=-30,vmax=30,transform=pc,cmap="RdBu_r")
axgr[1].coastlines()
axgr[1].set_xticks(np.linspace(-180, 180, 5), crs=pc)
axgr[1].set_yticks(np.linspace(-90, 90, 5), crs=pc)
axgr[1].set_title('transient - control (LR 2031-2050 av)')
axgr[1].tick_params(labelbottom=False,labelleft=False)

# p=axgr[2].pcolormesh(ds_t12.TLONG, ds_t12.TLAT, ssh_tmc1_HR,
#                        vmin=-30,vmax=30,transform=pc,cmap="RdBu_r")
p=axgr[2].pcolormesh(ssh_tmc1_rgd_HR.lon, ssh_tmc1_rgd_HR.lat, ssh_tmc1_rgd_HR,
                       vmin=-30,vmax=30,transform=pc,cmap="RdBu_r")
axgr[2].coastlines()
axgr[2].set_xticks(np.linspace(-180, 180, 5), crs=pc)
axgr[2].set_yticks(np.linspace(-90, 90, 5), crs=pc)
axgr[2].set_title('transient - control (HR 1951-1970 av)')


# p=axgr[3].pcolormesh(ds_t12.TLONG, ds_t12.TLAT, ssh_tmc2_HR,
#                        vmin=-30,vmax=30,transform=pc,cmap="RdBu_r")
p=axgr[3].pcolormesh(ssh_tmc2_rgd_HR.lon, ssh_tmc2_rgd_HR.lat, ssh_tmc2_rgd_HR,
                       vmin=-30,vmax=30,transform=pc,cmap="RdBu_r")
axgr[3].coastlines()
axgr[3].set_xticks(np.linspace(-180, 180, 5), crs=pc)
axgr[3].set_yticks(np.linspace(-90, 90, 5), crs=pc)
axgr[3].set_title('transient - control (HR 2031-2050 av)')
axgr[3].tick_params(labelleft=False)

cb = axgr.cbar_axes[0].colorbar(p)
cax = axgr.cbar_axes[0]
axis = cax.axis[cax.orientation]
axis.label.set_text("SSH [cm]")

plt.savefig(pic1,bbox_inches='tight')
```
