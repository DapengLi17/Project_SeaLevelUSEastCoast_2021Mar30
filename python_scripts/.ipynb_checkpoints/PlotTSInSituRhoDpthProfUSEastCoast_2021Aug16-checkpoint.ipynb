# data analysis 

```python
%%time
temp1_r_TL = ds1_TL.TEMP.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16),
                              nlat=slice(nlat1_r_g16, nlat2_r_g16))

temp2_r_TL = ds2_TL.TEMP.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16),
                              nlat=slice(nlat1_r_g16, nlat2_r_g16))

temp1_r_CL = ds1_CL.TEMP.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16),
                              nlat=slice(nlat1_r_g16, nlat2_r_g16))

temp2_r_CL = ds2_CL.TEMP.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16),
                              nlat=slice(nlat1_r_g16, nlat2_r_g16))


salt1_r_TL = ds1_TL.SALT.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16),
                              nlat=slice(nlat1_r_g16, nlat2_r_g16))

salt2_r_TL = ds2_TL.SALT.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16),
                              nlat=slice(nlat1_r_g16, nlat2_r_g16))

salt1_r_CL = ds1_CL.SALT.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16),
                              nlat=slice(nlat1_r_g16, nlat2_r_g16))

salt2_r_CL = ds2_CL.SALT.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16),
                              nlat=slice(nlat1_r_g16, nlat2_r_g16))

rho1_r_TL, drhodS1_r_TL, drhodT1_r_TL  = pop_tools.eos(salt1_r_TL, temp1_r_TL, depth=temp1_r_TL.z_t*1e-2, return_coefs=True)
rho2_r_TL = pop_tools.eos(salt2_r_TL, temp2_r_TL, depth=temp2_r_TL.z_t*1e-2)
rho1_r_CL = pop_tools.eos(salt1_r_CL, temp1_r_CL, depth=temp1_r_CL.z_t*1e-2)
rho2_r_CL = pop_tools.eos(salt2_r_CL, temp2_r_CL, depth=temp2_r_CL.z_t*1e-2)

beta1_r_TL = drhodS1_r_TL/rho1_r_TL/1000 
# https://github.com/NCAR/pop-tools/blob/master/pop_tools/eos.py DRHODS * 1000 (probably for unit convertion), 
# haline coefficient is ~ 1e-4 (e.g., Fig 1., Griffies et al., 2014 An assessment of global and regional sea level 
# for years 1993–2007 in a suite of interannual CORE-II simulations), so I divide 1000 to get the correct order of magnitude
alpha1_r_TL = -drhodT1_r_TL/rho1_r_TL

rho1_r_TL = rho1_r_TL.mean('time').compute()
rho2_r_TL = rho2_r_TL.mean('time').compute()

rho1_r_CL = rho1_r_CL.mean('time').compute()
rho2_r_CL = rho2_r_CL.mean('time').compute()

alpha1_r_TL = alpha1_r_TL.mean('time').compute()
beta1_r_TL = beta1_r_TL.mean('time').compute()

dtemp_r_TLmCL = ((temp2_r_TL.mean('time')-temp1_r_TL.mean('time')) - 
                 (temp2_r_CL.mean('time')-temp1_r_CL.mean('time'))).compute()
dsalt_r_TLmCL = ((salt2_r_TL.mean('time')-salt1_r_TL.mean('time')) - 
                 (salt2_r_CL.mean('time')-salt1_r_CL.mean('time'))).compute()
drho_r_TLmCL = (rho2_r_TL-rho1_r_TL) - (rho2_r_CL-rho1_r_CL)
drho_r_TLmCL
```

```python
%%time
# temp1_TH = xr.concat([ds1a_temp_TH.TEMP, ds1b_TH.TEMP],dim='time')
# salt1_TH = xr.concat([ds1a_salt_TH.SALT, ds1b_TH.SALT],dim='time')
# print(temp1_TH.time)

temp1_r_TH = ds1_temp_TH.TEMP.isel(nlon=slice(nlon1_r_t12, nlon2_r_t12),
                                   nlat=slice(nlat1_r_t12, nlat2_r_t12))

temp2_r_TH = ds2_temp_TH.TEMP.isel(nlon=slice(nlon1_r_t12, nlon2_r_t12),
                                   nlat=slice(nlat1_r_t12, nlat2_r_t12))

temp1_r_CH = ds1_temp_CH.TEMP.isel(nlon=slice(nlon1_r_t12, nlon2_r_t12),
                                   nlat=slice(nlat1_r_t12, nlat2_r_t12))

temp2_r_CH = ds2_temp_CH.TEMP.isel(nlon=slice(nlon1_r_t12, nlon2_r_t12),
                                   nlat=slice(nlat1_r_t12, nlat2_r_t12))


salt1_r_TH = ds1_salt_TH.SALT.isel(nlon=slice(nlon1_r_t12, nlon2_r_t12),
                              nlat=slice(nlat1_r_t12, nlat2_r_t12))

salt2_r_TH = ds2_salt_TH.SALT.isel(nlon=slice(nlon1_r_t12, nlon2_r_t12),
                                   nlat=slice(nlat1_r_t12, nlat2_r_t12))

salt1_r_CH = ds1_salt_CH.SALT.isel(nlon=slice(nlon1_r_t12, nlon2_r_t12),
                                   nlat=slice(nlat1_r_t12, nlat2_r_t12))

salt2_r_CH = ds2_salt_CH.SALT.isel(nlon=slice(nlon1_r_t12, nlon2_r_t12),
                                   nlat=slice(nlat1_r_t12, nlat2_r_t12))

rho1_r_TH, drhodS1_r_TH, drhodT1_r_TH = pop_tools.eos(salt1_r_TH, temp1_r_TH, depth=temp1_r_TH.z_t*1e-2, return_coefs=True)
rho2_r_TH = pop_tools.eos(salt2_r_TH, temp2_r_TH, depth=temp2_r_TH.z_t*1e-2)
rho1_r_CH = pop_tools.eos(salt1_r_CH, temp1_r_CH, depth=temp1_r_CH.z_t*1e-2)
rho2_r_CH = pop_tools.eos(salt2_r_CH, temp2_r_CH, depth=temp2_r_CH.z_t*1e-2)

beta1_r_TH = drhodS1_r_TH/rho1_r_TH/1000 
# https://github.com/NCAR/pop-tools/blob/master/pop_tools/eos.py DRHODS * 1000 (probably for unit convertion), 
# haline coefficient is ~ 1e-4 (e.g., Fig 1., Griffies et al., 2014 An assessment of global and regional sea level 
# for years 1993–2007 in a suite of interannual CORE-II simulations), so I divide 1000 to get the correct order of magnitude
alpha1_r_TH = -drhodT1_r_TH/rho1_r_TH


rho1_r_TH = rho1_r_TH.mean('time').compute()
rho2_r_TH = rho2_r_TH.mean('time').compute()

rho1_r_CH = rho1_r_CH.mean('time').compute()
rho2_r_CH = rho2_r_CH.mean('time').compute()

alpha1_r_TH = alpha1_r_TH.mean('time').compute()
beta1_r_TH = beta1_r_TH.mean('time').compute()

dtemp_r_THmCH = ((temp2_r_TH.mean('time')-temp1_r_TH.mean('time')) - 
                 (temp2_r_CH.mean('time')-temp1_r_CH.mean('time'))).compute()
dsalt_r_THmCH = ((salt2_r_TH.mean('time')-salt1_r_TH.mean('time')) - 
                 (salt2_r_CH.mean('time')-salt1_r_CH.mean('time'))).compute()
drho_r_THmCH = (rho2_r_TH-rho1_r_TH) - (rho2_r_CH-rho1_r_CH)
drho_r_THmCH
```
