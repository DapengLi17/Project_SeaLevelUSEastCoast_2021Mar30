```python
ds0 = xr.open_dataset(infile0,chunks=chunk_size)
ds0
```

```python
droplist = list(ds0.variables)
keeplist = ['SSH','z_t','z_w','z_w_top','dz','dzw','time','TLONG','TLAT','ULONG','ULAT','KMT','KMU','TAREA','UAREA']
for i in keeplist:
    droplist.remove(i)
droplist
```

```python
%%time
ds = xr.open_mfdataset(infiles, compat="override", combine="by_coords", data_vars="minimal",
                          coords="minimal", chunks=chunk_size, drop_variables=droplist, parallel=True)
ds
```

```python
%%time
outfile_ssh = outdir + prefix + '1950_2050DailySSH_' + date_str + '.nc'
ds.SSH.to_netcdf(path=outfile_ssh, mode='w', format='NETCDF4', compute=True)
ds.SSH
```

```python
%%time
outfile_temp = outdir + prefix + '1950_2050MonthlyTEMP_' + date_str + '.nc'
ds.TEMP.to_netcdf(path=outfile_ssh, mode='w', format='NETCDF4', compute=True)
ds.TEMP
```
