transient run (RCP85) - HR

```python
chunk_size = {'nlat':800,'nlon':900}
mons=np.arange(1,13)
yrs = np.arange(2080,2100+1)
indir = ('/scratch/user/dapengli/Projects4iHESP/Project_SeaLevelUSEastCoast_2021Mar30/'
'data_after_manipulation/RegionalStericHeight/test/B.E.13.BRCP85C5CN.ne120_t12.sehires38.003.sunway.CN_OFF/')
prefix= 'B.E.13.BRCP85C5CN.ne120_t12.sehires38.003.sunway.CN_OFF.pop.h._RegionalStericHeight_'
infiles = [indir + prefix + str(iyr) +'-'+str(imon).zfill(2)+ '_2021Apr15.nc' 
               for iyr, imon in product(yrs, mons)]
outfile = ('/scratch/user/dapengli/Projects4iHESP/Project_SeaLevelUSEastCoast_2021Mar30/'
           'data_after_manipulation/RegionalStericHeight/test_2021Apr20/'
           'B.E.13.BRCP85C5CN.ne120_t12.sehires38.003.sunway.CN_OFF/' + prefix + '2080-2100_2021Apr24.nc')
```

# data data and data analysis 

```python
ds = xr.open_mfdataset(infiles, compat="override", combine="by_coords", data_vars="minimal",
                          coords="minimal", chunks=chunk_size, parallel=True)
time_s = ds.time.to_pandas().index.shift(-1,'M') 
ds = ds.assign_coords(time=time_s)
ds
```

```python
h_rst_yr = ds.h_rst.resample(time='A').mean('time')
h_rst_yr
```

# save outputs

```python
%%time
h_rst_yr.to_netcdf(path=outfile, mode='w', format='NETCDF4', compute=True)
```
