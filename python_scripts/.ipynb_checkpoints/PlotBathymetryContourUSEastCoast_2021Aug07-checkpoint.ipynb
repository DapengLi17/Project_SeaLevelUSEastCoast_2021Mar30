# define parameters and load files

```python
chunks_HR = {'nlat':800,'nlon':900}
chunks_LR = {'nlat':384,'nlon':320}


# === Gulf of Maine (GMa) ===
# lat1_GMa, lat2_GMa = 35, 55
# lon1_GMa, lon2_GMa = -75, -50 # GMa: Gulf of Maine

# lon1_GMa, lon2_GMa = -73, -63 
# lat1_GMa, lat2_GMa = 39, 47
lon1_GMa, lon2_GMa = -73, -63
lat1_GMa, lat2_GMa = 39, 47

lon1_bx1, lon2_bx1 = -71, -66  # b: box, highlight region with a box 
lat1_bx1, lat2_bx1 = 42, 45 
# ===========================


# === GMa: Gulf of Maine ===
# lat1_GMa, lat2_GMa = 35, 55
# lon1_GMa, lon2_GMa = -75, -50 
# ==========================

# === GMe: Gulf of Mexico ===
# lat1_GMe, lat2_GMe = 15, 32
# lon1_GMe, lon2_GMe = -100, -75 
# ===========================


# === r: region (US east coast excluding Gulf of Mexico) ===
# reference: Plot2FindGulfMexicoYucatanChannelFLstraitnlatnlonIndex_2021Jul20.ipynb
nlon1_r_t12, nlon2_r_t12 = 200, 700 
nlat1_r_t12, nlat2_r_t12 = 1300, 2000

nlon1_r_g16, nlon2_r_g16 = 260, 321
nlat1_r_g16, nlat2_r_g16 = 230, 350

lon1_r, lon2_r = -85, -50
lat1_r, lat2_r = 25, 55
# ===========================================================


outdir='/scratch/user/dapengli/Projects4iHESP/Project_SeaLevelUSEastCoast_2021Mar30/pics/'
pic1 = outdir + 'BathymetryContourUSEastCoast_2021Aug07.png'
```

# read files

```python
infile_etopo = '/ihesp/user/dapengli/obs_dat/bathymetry_ETOPO5_2021Apr27/data.nc'
ds_etopo = xr.open_dataset(infile_etopo)
ds_etopo

ds_etopo_rn = ds_etopo.rename({'X': 'lon', 'Y': 'lat'}) # rn: rename
ds_etopo_rn
```

```python
# supplimentary files 
infile0_LR = ('/ihesp/user/dapengli/Projects4iHESP/SupplementaryFiles4POP/'
              'cmpr_B.E.13.BHISTC5.ne30g16.sehires38.003.sunway.pop.h.1850-01.nc')
ds0_LR = xr.open_dataset(infile0_LR)
ds0_r_LR = ds0_LR.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16),nlat=slice(nlat1_r_g16, nlat2_r_g16)) 

infile0_HR = ('/ihesp/user/dapengli/Projects4iHESP/SupplementaryFiles4POP/'
              'B.E.13.BHISTC5.ne120_t12.sehires38.003.sunway.pop.h.1850-01.nc')
ds0_HR = xr.open_dataset(infile0_HR)
ds0_r_HR = ds0_HR.isel(nlon=slice(nlon1_r_t12, nlon2_r_t12),nlat=slice(nlat1_r_t12, nlat2_r_t12)) 

infile_g16 = ('/ihesp/user/dapengli/Projects4iHESP/'
              'SupplementaryFiles4POP/POPgrids_gx1v6_2020Jun29.nc') # POP g16 grid file
ds_g16 = xr.open_dataset(infile_g16)
ds_r_g16 = ds_g16.isel(nlon=slice(nlon1_r_g16, nlon2_r_g16),nlat=slice(nlat1_r_g16, nlat2_r_g16))


infile_t12 = ('/ihesp/user/dapengli/Projects4iHESP/'
              'SupplementaryFiles4POP/POPgrids_tx0.1v2_2020Jun29.nc') # POP t12 grid file
ds_t12 = xr.open_dataset(infile_t12)
ds_r_t12 = ds_t12.isel(nlon=slice(nlon1_r_t12, nlon2_r_t12),nlat=slice(nlat1_r_t12, nlat2_r_t12))
```

# data analysis

```python
bathy_etopo=ds_etopo_rn.elev
bathy_etopo=-bathy_etopo.where(ds_etopo_rn.elev<0,np.nan) 
# mask land topo as nan, make bathy positive values

bathy_r_etopo = bathy_etopo.sel(lon=slice(lon1_r+360,lon2_r+360),
                                 lat=slice(lat2_r,lat1_r))

fig, axs = plt.subplots(1,2,figsize=(15,5))
ds_etopo.elev.plot(ax=axs[0])
bathy_r_etopo.plot(ax=axs[1],vmin=0,vmax=5000)
bathy_etopo
```

# make pics

```python
pc = cartopy.crs.PlateCarree()
lon_formatter = LongitudeFormatter(zero_direction_label=True)
lat_formatter = LatitudeFormatter()

fig1, axs = plot.subplots(nrows=3,ncols=2,proj='cyl',width='7.5in',height='10in')

m00=axs[0,0].contourf(bathy_r_etopo.lon, bathy_r_etopo.lat, 
bathy_r_etopo, levels=np.arange(0,500+1,50), 
        transform=pc, vmin=0, vmax=500, cmap="spectral", extend='max')# rdylbu
axs[0,0].coastlines(resolution='50m', color='black', linewidth=1)
# axs[0,0].format(land=True, landcolor='grey')
axs[0,0].set_extent([lon1_r, lon2_r, lat1_r, lat2_r], crs=pc)
axs[0,0].set_xticks(np.arange(lon1_r,lon2_r+1,10), crs=pc)
axs[0,0].set_xticklabels(np.arange(lon1_r,lon2_r+1,10), fontsize=12)
axs[0,0].xaxis.set_major_formatter(lon_formatter)
axs[0,0].set_yticks(np.arange(lat1_r,lat2_r+1, 10), crs=pc)
axs[0,0].set_yticklabels(np.arange(lat1_r,lat2_r+1, 10), fontsize=12)
axs[0,0].yaxis.set_major_formatter(lat_formatter)
axs[0,0].set_title('Bathymetry ETOPO5', fontsize=15, fontweight='bold')
axs[0,0].text(-84, 53, 'a', transform=pc, fontsize=10)

              
m10=axs[1,0].contourf(ds_r_t12.TLONG, ds_r_t12.TLAT, ds0_r_HR.HT.where(ds0_r_HR.KMT>0,np.nan)/100,
        levels=np.arange(0,500+1,50),
        transform=pc, vmin=0, vmax=500, cmap="spectral", extend='max')
axs[1,0].coastlines(resolution='50m', color='black', linewidth=1)
axs[1,0].set_extent([lon1_r, lon2_r, lat1_r, lat2_r], crs=pc)
axs[1,0].set_xticks(np.arange(lon1_r,lon2_r+1,10), crs=pc)
axs[1,0].set_xticklabels(np.arange(lon1_r,lon2_r+1,10), fontsize=12)
axs[1,0].xaxis.set_major_formatter(lon_formatter)
axs[1,0].set_yticks(np.arange(lat1_r,lat2_r+1, 10), crs=pc)
axs[1,0].set_yticklabels(np.arange(lat1_r,lat2_r+1, 10), fontsize=12)
axs[1,0].yaxis.set_major_formatter(lat_formatter)
axs[1,0].set_title('Bathymetry HR', fontsize=15, fontweight='bold')
axs[1,0].text(-84, 53, 'b', transform=pc, fontsize=10)


m20=axs[2,0].contourf(ds_r_g16.TLONG, ds_r_g16.TLAT, ds0_r_LR.HT.where(ds0_r_LR.KMT>0,np.nan)/100,
        levels=np.arange(0,500+1,50),
        transform=pc, vmin=0, vmax=500, cmap="spectral", extend='max')
axs[2,0].coastlines(resolution='50m', color='black', linewidth=1)
axs[2,0].set_extent([lon1_r, lon2_r, lat1_r, lat2_r], crs=pc)
axs[2,0].set_xticks(np.arange(lon1_r,lon2_r+1,10), crs=pc)
axs[2,0].set_xticklabels(np.arange(lon1_r,lon2_r+1,10), fontsize=12)
axs[2,0].xaxis.set_major_formatter(lon_formatter)
axs[2,0].set_yticks(np.arange(lat1_r,lat2_r+1, 10), crs=pc)
axs[2,0].set_yticklabels(np.arange(lat1_r,lat2_r+1, 10), fontsize=12)
axs[2,0].yaxis.set_major_formatter(lat_formatter)
axs[2,0].set_title('Bathymetry LR', fontsize=15, fontweight='bold')
axs[2,0].text(-84, 53, 'c', transform=pc, fontsize=10)
# add colorbar
cb = axs[2,0].colorbar(m20, loc='b', label='[m]', extend='max',width='2em',shrink=0.9)

m01=axs[0,1].pcolormesh(bathy_r_etopo.lon, bathy_r_etopo.lat, 
        bathy_r_etopo, levels=np.arange(0,500+1,50), 
        transform=pc, vmin=0, vmax=500, cmap="spectral", extend='max')# rdylbu
axs[0,1].coastlines(resolution='50m', color='black', linewidth=1)
axs[0,1].set_extent([lon1_GMa, lon2_GMa, lat1_GMa, lat2_GMa], crs=pc)
axs[0,1].set_xticks(np.arange(lon1_GMa,lon2_GMa+1,2), crs=pc)
axs[0,1].set_xticklabels(np.arange(lon1_GMa,lon2_GMa+1,2), fontsize=12)
axs[0,1].xaxis.set_major_formatter(lon_formatter)
axs[0,1].set_yticks(np.arange(lat1_GMa,lat2_GMa+1, 2), crs=pc)
axs[0,1].set_yticklabels(np.arange(lat1_GMa,lat2_GMa+1, 2), fontsize=12)
axs[0,1].yaxis.set_major_formatter(lat_formatter)
axs[0,1].set_title('Bathymetry ETOPO5', fontsize=15, fontweight='bold')
axs[0,1].text(-72.7, 46.5, 'd', transform=pc, fontsize=10)
axs[0,1].text(-65.9, 42.1+0.2, 'NC', transform=pc, fontsize=8) # Northeast Channel
axs[0,1].scatter(-68+0.05, 43.5, c='k', s=10, transform=pc)
axs[0,1].text(-68+0.05-0.2, 43.5+0.2, 'JB', transform=pc, fontsize=8) # Jordan Basin
axs[0,1].scatter(-67, 42.5, c='k', s=10, transform=pc)
axs[0,1].text(-67-0.2, 42.5+0.2, 'GB', transform=pc, fontsize=8) # Georges Basin
axs[0,1].scatter(-65.9, 42.1, c='k', s=10, transform=pc)
axs[0,1].text(-70+0.5-0.2, 42.5+0.2, 'WB', transform=pc, fontsize=8) # Wilkinson Basin
axs[0,1].scatter(-70+0.5, 42.5, c='k', s=10, transform=pc)
ap=axs[0,1].add_patch(mpatches.Rectangle(xy=[lon1_bx1, lat1_bx1], 
                                         width=(lon2_bx1-lon1_bx1), height=(lat2_bx1-lat1_bx1),
                                         facecolor='none',transform=pc,
                                         linewidth=1, edgecolor='k'))
ap.set_zorder(2)

m11=axs[1,1].pcolormesh(ds_r_t12.TLONG, ds_r_t12.TLAT, ds0_r_HR.HT.where(ds0_r_HR.KMT>0,np.nan)/100,
        levels=np.arange(0,500+1,50),
        transform=pc, cmap="spectral", extend='max')
axs[1,1].coastlines(resolution='50m', color='black', linewidth=1)
axs[1,1].set_extent([lon1_GMa, lon2_GMa, lat1_GMa, lat2_GMa], crs=pc)
axs[1,1].set_xticks(np.arange(lon1_GMa,lon2_GMa+1,2), crs=pc)
axs[1,1].set_xticklabels(np.arange(lon1_GMa,lon2_GMa+1,2), fontsize=12)
axs[1,1].xaxis.set_major_formatter(lon_formatter)
axs[1,1].set_yticks(np.arange(lat1_GMa,lat2_GMa+1, 2), crs=pc)
axs[1,1].set_yticklabels(np.arange(lat1_GMa,lat2_GMa+1, 2), fontsize=12)
axs[1,1].yaxis.set_major_formatter(lat_formatter)
axs[1,1].set_title('Bathymetry HR', fontsize=15, fontweight='bold')
axs[1,1].text(-72.7, 46.5, 'e', transform=pc, fontsize=10)

m21=axs[2,1].pcolormesh(ds_r_g16.TLONG, ds_r_g16.TLAT, ds0_r_LR.HT.where(ds0_r_LR.KMT>0,np.nan)/100,
        levels=np.arange(0,500+1,50),
        transform=pc, cmap="spectral", extend='max')
axs[2,1].coastlines(resolution='50m', color='black', linewidth=1)
axs[2,1].set_extent([lon1_GMa, lon2_GMa, lat1_GMa, lat2_GMa], crs=pc)
axs[2,1].set_xticks(np.arange(lon1_GMa,lon2_GMa+1,2), crs=pc)
axs[2,1].set_xticklabels(np.arange(lon1_GMa,lon2_GMa+1,2), fontsize=12)
axs[2,1].xaxis.set_major_formatter(lon_formatter)
axs[2,1].set_yticks(np.arange(lat1_GMa,lat2_GMa+1, 2), crs=pc)
axs[2,1].set_yticklabels(np.arange(lat1_GMa,lat2_GMa+1, 2), fontsize=12)
axs[2,1].yaxis.set_major_formatter(lat_formatter)
axs[2,1].set_title('Bathymetry LR', fontsize=15, fontweight='bold')
axs[2,1].text(-72.7, 46.5, 'f', transform=pc, fontsize=10)
# add colorbar
cb = axs[2,1].colorbar(m21, loc='b', label='[m]', extend='max',width='2em',shrink=0.9)

fig1.savefig(pic1,bbox_inches='tight')
```
