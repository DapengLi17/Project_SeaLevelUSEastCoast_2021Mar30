    true_zonal_velocity = cos(ANGLE) * UVEL - sin(ANGLE) * VVEL  <br>
    true_merid_velocity = sin(ANGLE) * UVEL + cos(ANGLE) * VVEL  <br>
    reference: https://www.ncl.ucar.edu/Applications/popvec.shtml<br> 

```python
u1r_NA_TL = np.cos(angle_NA_TL)*u1_NA_TL - np.sin(angle_NA_TL)*v1_NA_TL 
v1r_NA_TL = np.sin(angle_NA_TL)*u1_NA_TL + np.cos(angle_NA_TL)*v1_NA_TL 

dur_NA_TLmCL = np.cos(angle_NA_TL)*du_NA_TLmCL - np.sin(angle_NA_TL)*dv_NA_TLmCL 
dvr_NA_TLmCL = np.sin(angle_NA_TL)*du_NA_TLmCL + np.cos(angle_NA_TL)*dv_NA_TLmCL 
```
